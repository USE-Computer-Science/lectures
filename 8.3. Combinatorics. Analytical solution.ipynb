### **Аналитическое решение комбинаторики**

```python
from itertools import product
from itertools import permutations
from itertools import combinations
from itertools import combinations_with_replacement
```

#### **1. Размещения с повторениями**

Число размещений с повторениями из множества ***n*** по ***k*** элементов:
<br>
<br>
$$\overline{A^k_n} = n^k$$

**Пример 1.** 
<br>
Пусть, требуется посчитать количество всех возможных двухбуквенных слов,  составленных из букв слова **ГОД**.
<br>
<br>
$$\overline{A^2_3} = 3^2 = 9$$

image.png

```python
# Число размещений с повторениями из алфавита 'год' по 2 элемена (декартово произведение)
# n = len('год')
# k = repeat

# все возможные размещения
print([*product('год', repeat = 2)])

# количество размещений
print(len([*product('год', repeat = 2)]))
```

#### **2. Размещения без повторений**

Число размещений без повторений из множества ***n*** по ***k*** элементов:
<br>
<br>
$$A^k_n = \frac{n!}{(n-k)!}$$

**Пример 2.** 
<br>
Пусть, требуется подсчитать количество всех возможных двухбуквенных слов, составленных из букв слова **ГОД**, но буквы в полученных словах не должны повторяться.
<br>
<br>
$$A^2_3 = \frac{3!}{(3-2)!}= 3 \cdot 2 = 6$$

image.png

```python
# Число размещений без повторений из алфавита 'год' по 2 элемента
# n = len('год')
# k = r

# все возможные размещения без повторений
print([*permutations('год', r = 2)])

# количество размещений без повторений
print(len([*permutations('год', r = 2)]))
```

#### **3. Сочетания**

Число сочетаний (размещений без учёта порядка) из множества ***n*** по ***k*** элементов:
<br>
<br>
$$C^k_n = \frac{A^k_n}{k!} = \frac{n!}{k! \, (n-k)!}$$

**Пример 3.** 
<br>
Сколько двухбуквенных слов в возрастающем алфавитном порядке можно составить из букв слова **ГОД**.
<br>
<br>
$$C^2_3 = \frac{3!}{2! \, (3-2)!} = \frac{3!}{2!} = 3$$

image.png

```python
# число сочетаний из алфавита 'год' по 2 элемента (число размещений без повторений и учёта порядка)
# n = len('год')
# k = r

# все возможные сочетания
print([*combinations('год', r = 2)])

# количество сочетаний
print(len([*combinations('год', r = 2)]))
```

#### **4. Сочетания с повторениями**

Число сочетаний с повторениями из множества ***n*** по ***k*** элементов:
<br>
<br>
$$\overline{C^k_n} = C^k_{n+k-1} =  \frac{A^k_{n+k-1}}{k!} = \frac{(n + k - 1)!}{k! \; (n-1)!}$$

**Пример 4.** 
<br>
Сколько двухбуквенных слов в неубывающем алфавитном порядке можно составить из букв слова **ГОД**.
<br>
<br>
$$\overline{C^2_3} = \frac{(3 + 2 - 1)!}{2! \, (3-1)!} = \frac{4!}{2! \, 2!} = \frac{4 \cdot 3}{2} = 6$$

image.png

```python
# число сочетаний с повторениями из алфавита 'год' по 2 элемента (число размещений без учёта порядка с повторениями)
# n = len('год')
# k = r

# все возможные сочетания с повторениями
print([*combinations_with_replacement('год', r = 2)])

# количество сочетаний с повторениями
print(len([*combinations_with_replacement('год', r = 2)]))
```

image.png

На первом месте числа могут стоять только цифры 1, 3, 5, 7, 8 = 5 цифр.

На оставшихся шести позициях могут стоят любые цифры $9 \cdot 9 \cdot 9 \cdot 9 \cdot 9 \cdot 9 = 9^6$
<br>
Всего допустимых комбинаций $5 \cdot 9^6$
<br>
Из этих комбинаций остаётся извлечь комбинации с тройкой одинаковых цифр

Таких комбинаций $5 \cdot 9 \cdot 9 \cdot 9 \cdot 9 = \cdot 9^4$
Итого можно получить $5 \cdot 9^6 - 5 \cdot 9^4 = 2 624 400$ чисел

#####***(№ 5)*** Из букв слова КАНКАН составляются 6-буквенные последовательности. Сколько можно составить различных последовательностей, если известно, что в каждой из них содержится не менее 3 согласных?

Ответ: 656

image.png

$$U = \overline{A^6_3} = 3^6 = 729$$
<br>
$$ {6_A} = C^6_6 = 1$$
$$ {5_A} = C^5_6 \cdot \overline{A^1_2} = 6 \cdot 2 = 12$$
$$ {4_A} = C^4_6 \cdot \overline{A^2_2} = 15 \cdot 2^2 = 60$$
<br>
$$U - ({6_A} + {5_A} + {4_A}) = U - (C^6_6 + C^5_6 \cdot \overline{A^1_2} + C^4_6 \cdot \overline{A^2_2}) = 729 - 73 = 656 $$

```python
U = len([*product('КАН',repeat=6)])
A6 = len([*combinations('123456',r=6)])
A5 = len([*combinations('123456',r=5)]) * len([*product('КН',repeat=1)])
A4 = len([*combinations('123456',r=4)]) * len([*product('КН',repeat=2)])
print(U - A6 - A5 - A4)
```
