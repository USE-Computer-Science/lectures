#### **Задачи на траекторию**

**Пример 1.** Исполнитель преобразовывает число на экране. У исполнителя есть две команды, которым присвоены номера:

1. Прибавить 1
2. Умножить на 2

При выполнении первой команды число на экране увеличивается на 1, при выполнении второй – увеличивается в 2 раза. Программа для исполнителя – это последовательность команд. Сколько существует программ, которые число 1 преобразовывают в число 12?

Ответ: 20

```python
# рекурсивная функция обхода дерева вычислений
# cur - текущее значение на экране
# fin - число, которое необходимо достичь
def f(cur, fin):
  # если текущее число больше числа назначения - нет ни одной траектории (т.к. все команды увеличивают cur)
  if cur > fin: return 0
  # если текущее число равно числу назначения - одна из траекторий найдена (т.к. мы попали в fin)
  if cur == fin: return 1
  # если текущее число меньше числа назначения - продолжаем рекурсивный подсчёт суммы путей
  return f(cur+1, fin) + f(cur*2, fin)

# печатаем ответ
print(f(1, 12))
```

Выполняя первую команду, исполнитель увеличивает текущее число на 1, а выполняя вторую – из числа x получает число 2x+1. Программой для исполнителя называется последовательность команд. Траектория вычислений — это последовательность результатов выполнения всех команд программы. Например, для программы «121» при исходном числе 1 траектория будет состоять из чисел 1+1=2, 2*2+1=5, 5+1=6. Сколько существует программ, для которых при исходном числе 1 результатом является число 31 и при этом траектория вычислений не содержит число 25?

Ответ: 3875

```python
# рекурсивная функция обхода дерева вычислений
# cur - текущее значение на экране
# fin - число, которое необходимо достичь
def f(cur, fin):
  # исключаем траектории, не попадающие в пункт назначения
  if cur > fin: return 0
  # если текущее число равно числу назначения - одна из траекторий найдена (т.к. мы попали в fin)
  if cur == fin: return 1
  # если текущее число меньше числа назначения - продолжаем рекурсивный подсчёт суммы путей
  return f(cur+1, fin) + f(cur+2, fin)

# событие A - количество траекторий, проходящих через мост в 15
# событие B - количество траекторий, проходящих через мост в 25
# A ∧ B - количество траекторий, проходящих через оба моста одновременно
# A ∨ B = A ∨ B - (A ∧ B)
f(11, 15) * f(15, 29) + f(11,25)*f(25,29) - f(11,15)*f(15,25)*f(25,29)
```

**Пример 1.** Исполнитель преобразовывает число на экране. У исполнителя есть три команды, которым присвоены номера:

1. Прибавить 1
2. Умножить на 2
3. Умножить на 3

Первая команда увеличивает число на 1, вторая – умножает его на 2, третья – умножает на 3. Программа для исполнителя – это последовательность команд. Сколько существует программ, состоящих ровно из 10 команд, для которых при исходном числе 2 результатом является число 42?

Ответ: 21

```python
# рекурсивная функция обхода дерева вычислений
# cur - текущее число на экране
# fin - число, которое необходимо получить
# step - номер текущей команды в программе (текущий уровень рекурсии)
def f(cur, fin, step):
  # если число fin получено ровно за 8 команд - одна из программ найдена  
  if step == 10 and cur == fin: return 1
  # если за 8 команд не удалось получить fin - обрываем эту ветвь рекурсии
  if step >= 10: return 0
  # если выполнено менее восьми команд - продолжаем рекурсивный обход
  # при увеличении уровня рекурсии должен увеличиваться и номер текущей команды
  # обеспечим это с помощью увеличения параметра step (step = step+1)  
  return f(cur+1,fin,step+1) + f(cur*2,fin,step+1) + f(cur*3,fin,step+1)

# печатаем ответ 
f(2,42,0)
```

**Пример 2.** Исполнитель преобразовывает число, записанное на экране. У него есть три команды, которым присвоены номера:

1. Прибавить 1, если число нечётное
2. Умножить на 2, если число чётное
3. Вычесть 3

Первая команда увеличивает число на экране на 1 и может использоваться только в том случае, если текущее число на экране нечётное. Вторая команда увеличивает число в 2 раза и доступна только для чётных текущих чисел. Третья команда уменьшает число на 3 и применима к любым числам. Программой для исполнителя называется последовательность команд. Сколько существует программ длиной не более 15 команд, которые преобразовывают число 5 в число 44?

Ответ: 150

```python
# рекурсивная функция обхода дерева вычислений
# cur - текущее число на экране
# fin - число, которое необходимо получить
# step - номер текущей команды в программе (уровень рекурсии)
def f(cur, fin, step):
  # если число fin получено не более чем за 8 команд - одна из программ найдена 
  if step <= 15 and cur == fin: return 1
  # если за 15 команд не удалось получить fin - обрываем эту ветвь рекурсии 
  if step >= 15: return 0
  # если выполнено менее пятнадцати команд - продолжаем рекурсивный обход 
  if cur % 2 != 0:
    # для нечётных чисел доступны операции сложения и вычитания
    return f(cur+1,fin,step+1) + f(cur-3,fin,step+1)
  if cur % 2 == 0:
    # для чётных чисел доступны операции умножения и вычитания
    return f(cur*2,fin,step+1) + f(cur-3,fin,step+1)

# печатаем ответ
f(5,44,0)
```

**Пример 3.**  Исполнитель преобразовывает число, записанное на экране. У него есть три команды, которым присвоены номера:

1. Прибавить 1
2. Прибавить 4
2. Умножить на 3

Первая команда увеличивает число на 1, вторая – на 4, третья – увеличивает число в 3 раза. Программой для исполнителя называется последовательность команд. Найдите длину самой короткой (по количеству команд) программы, которая преобразовывает число 7 в число 77.

Ответ: 5

```python
from functools import lru_cache
# используем кеширование для ускорения работы программы
@lru_cache()
# рекурсивная функция обхода дерева вычислений
def f(cur, fin, step):
  # если достигнуто число назначения - возвращаем затраченное количество команд 
  if cur == fin: return step
  # если число назначение превышено и достичь его не удалось - возвращаем бесконечность
  if cur > fin: return float('inf')
  # если число назначения ещё не достигнуто - продолжаем обход дерева вычислений
  # т.к. требуется найти самую короткую программу - используем функцию min
  return min( f(cur+1,fin,step+1), f(cur+4,fin,step+1), f(cur*3,fin,step+1) )

f(7,77,0)
```

**Пример 4.** Исполнитель преобразует число, записанное на экране. У него есть две команды:

1. Вычесть 1 
2. Умножить на -2 
3. Извлечь корень

Первая команда уменьшает число на 1, вторая – умножает его на -2, третья команда извлекает из числа квадратный корень, если число является квадратом целого числа. Сколько различных неотрицательных результатов можно получить из исходного числа 4096 в ходе исполнения программы, содержащей ровно 10 команд?

Ответ: 454

```python
# множество для хранения неотрицательных результатов исполнения программы
results = set()
# рекурсивная функция обхода дерева вычислений
def f(cur, step):
  # если выполнено 10 команд
  if step == 10:
    # и в результате получилось неотрицательное число
    if cur > 0:
      # добавляем его в множество
      results.add(cur)
    # завершаем обход дерева
    return
  # если 10 команд ещё не выполнено - продолжаем рекурсивных обход дерева
  f(cur-1,step+1)
  f(-2*cur,step+1)
  # корень извлекаем только если текущее значение > 0 и из него извлекается целый корень 
  if cur > 0 and (cur**0.5).is_integer(): f(cur**0.5,step+1)

# вызываем функцию обхода дерева
f(4096,0)
# печатаем ответ
print(len(results))
```

**Пример 5.** Исполнитель преобразовывает число, записанное на экране. У него есть три команды, которым присвоены номера:

1. Прибавить 1
2. Умножить на 2
3. Умножить на 3

Первая команда увеличивает число на экране на 1, вторая – умножает его на 2, третья – умножает на 3. Программой для исполнителя называется последовательность команд. Сколько существует различных программ, которые преобразовывают исходное число 5 в число 67 и не содержат двух подряд идущих команд умножения?

Ответ: 275

```python
# рекурсивная функция обхода дерева вычислений
# cur - текущее число на экране
# fin - число, которое необходимо получить
# id - идентификтор команды ('+','*') на предыдущем уровне рекурсии
def f(cur, fin, id):
  # если число fin достигнуто - одна из траекторий найдена
  if cur == fin: return 1
  # если число fin превышено - обрываем текущую ветвь рекурсии
  if cur > fin: return 0
  # если на предыдущем шаге использовалась команда умножения
  if id == '*':
    # на текущем шаге мы можем использовать только команду сложения
    return f(cur+2,fin,'+') # '+' - id операции на текущем шаге рекурсии
  # если же на предыдущем шаге команда умножения не использовалась
  else:
    # мы имеем право использовать любую из трёх команд
    return f(cur+1,fin,'+') + f(cur*2,fin,'*') + f(cur*3,fin,'*')

# в начале ни одна из команд не вызвана, поэтому id = ''
# печатаем ответ
f(5,67,'')
```
