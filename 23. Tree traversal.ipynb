```python
#     
# cur -    
# fin - ,   
# step -      (  )
def f(cur, fin, step):
  #   fin    8  -      
  if step == 10 and cur == fin: return 1
  #   8     fin -    
  if step >= 10: return 0
  #      -   
  #          
  #       step (step = step+1)  
  return f(cur+1,fin,step+1) + f(cur*2,fin,step+1) + f(cur*3,fin,step+1)

#   
f(2,42,0)
```

** 3.**    ,   .     ,   :

     1,    4,      3 .      .     (  ) ,    7   77.

: 5

```python
from functools import lru_cache
#      
@lru_cache()
#     
def f(cur, fin, step):
  #     -     
  if cur == fin: return step
  #          -  
  if cur > fin: return float('inf')
  #       -    
  # ..      -   min
  return min( f(cur+1,fin,step+1), f(cur+4,fin,step+1), f(cur*3,fin,step+1) )

f(7,77,0)
```

```python
#     
# cur -    
# fin - ,   
# id -   ('+','*')    
def f(cur, fin, id):
  #   fin  -    
  if cur == fin: return 1
  #   fin  -    
  if cur > fin: return 0
  #       
  if id == '*':
    #         
    return f(cur+2,fin,'+') # '+' - id     
  #         
  else:
    #        
    return f(cur+1,fin,'+') + f(cur*2,fin,'*') + f(cur*3,fin,'*')

#        ,  id = ''
#  
f(5,67,'')
```
