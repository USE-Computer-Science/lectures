Найдите разность между средним арифметическим и минимальным значениями температуры во второй половине дня (с 12:00). В ответе запишите только целую часть получившегося числа.

Ответ: 13

```python
# создаём поток для чтения файла 
with open('9-0.csv', newline='') as f:
  # пропускаем строку заголовков
  next(f)

  # инициализируем счётчик дней
  n = 0
  # инициализируем стреднее арифметическое измерений
  mean_temp = 0
  # инициализируем минимальное измерение
  min_temp = float('inf')

  # построчно считываем файл
  for s in f:
    # увеличиваем счётчик дней
    n += 1
    # считываем показания второй половины дня (с 13 ячейки) в список, разбивая строку по разделителю ';'
    row = [float(item.replace(',','.')) for item in s.split(';')[13:]]
    # считаем среднее арифметическое дня
    mean_temp += sum(row) / len(row)
    # находим минимальное измерение за всё время
    for item in row:
      if item < min_temp:
        min_temp = item
  # печатаем разность среднего средних арифметических дня с минимальным измерением 
  print(mean_temp / n - min_temp)
```

Сколько раз встречалась температура, равная округленному до десятых среднему арифметическому значению всех измерений в таблице?

Ответ: 14

```python
# счётчик дней
n = 0
# счётчик подходящих температур
cnt = 0
# инициализируем среднее арифметическое измерений
mean_temp = 0

# создаём поток для чтения файла 
with open('9-0.csv', newline='') as f:
  # пропускаем строку заголовков
  next(f)

  # построчно считываем файл
  for s in f:
    # увеличиваем счётчик дней
    n += 1
    # считываем показания второй половины дня (с 13 ячейки) в список, разбивая строку по разделителю ';'
    row = [float(item.replace(',','.')) for item in s.split(';')[1:]]
    # считаем сумму средних арифметических всех дней
    mean_temp += sum(row) / len(row)
  
  # считаем среднее арифметическое всех дней, округлённое до одного знака
  mean_temp = round((mean_temp / n), 1)
  # печатаем найденное среднее арифметическое
  print(f'mean = {mean_temp}')

# снова создаём поток для чтения файла 
with open('9-0.csv', newline='') as f:
  next(f)
  for s in f:
    row = [float(item.replace(',','.')) for item in s.split(';')[1:]]
    # обходим все наблюдения температуры
    for item in row:
      # если наблюдение совпало со средним арифметическим 
      if item == mean_temp:
        # увеличиваем счётчик
        cnt += 1

  # печатаем ответ
  print(f'count = {cnt}')
```

**1.** В файле электронной таблицы в каждой строке содержатся четыре натуральных числа. Определите количество строк таблицы, для которых выполнены оба условия:

В строке есть одно повторяющееся ровно 2 раза число;

Сумма квадратов повторяющихся чисел меньше квадрата суммы неповторяющихся.

Ответ: 116

```python
# инициализируем счётчик
cnt = 0

# построчно считываем файл
for s in open('9-24.csv'):
  # разбиваем строку в список с помощью метода split('разделитель')
  # преобразовываем каждый элемент получаемого списка к целому числу int(item)
  row = [int(item) for item in s.split(';')]
  
  # печатаем для отладки строки файла
  #print(row)

  # пропускаем строки, где не ровно 1 повторяющийся 2 раза элемент
  if len(row) != len(set(row)) + 1:
    continue
  
  # сумма квадратов повторяющихся чисел
  dup_square = (sum(row) - sum(set(row)) )**2 + (sum(row) - sum(set(row)) )**2
  # квадрат суммы неповторяющихся
  unique_square = ( sum(row) - (sum(row) - sum(set(row)))*2 )**2

  # если сумма квадратов повторяющихся < квадрата суммы неповторяющихся
  if  dup_square < unique_square:
    # увеличиваем счётчик
    cnt += 1

# печатаем ответ
print(cnt)
```

1. удвоенный квадрат минимального числа больше, чем произведение двух других чисел, ни одно из которых не равно максимальному;

**6.** В каждой строке файла электронной таблицы содержатся пять натуральных чисел. Определите количество строк таблицы, для которых выполнены три условия:

1. В строке есть число повторяющееся ровно 3 раза.

2. В строке есть хотя бы одно неповторяющееся число.

2. Наибольшее из пяти чисел строки больше суммы повторяющихся.

Ответ: 3

```python
# инициализируем счётчик 
cnt = 0

# построчно считываем файл
for s in open('9-25.csv'):

  # получаем список из строки
  row = [int(item) for item in s.split(';')] 
  
  # сортируем строку
  row = sorted(row)

  # инициализируем повторяющееся число
  triple = 0
  
  # признак наличия трижды повторяющегося числа
  ok = False
  
  #  поэлементно обходим строку
  for item in row:
    # проверяем не нашлось ли число, повторяющееся 3 раза
    if row.count(item) == 3:
      triple = item
      ok = True
  
  # пропускаем строки, в которых нет трижды повторяющегося числа 
  if not ok:
    continue

  # если наибольшее число в строке больше суммы повторяющихся
  if max(row) > 3 * triple:
    # увеличиваем счётчик
    cnt += 1

# печатаем ответ
print(cnt)
```

**8.** В файле электронной таблицы в каждой строке содержатся шесть натуральных чисел. Определите количество строк таблицы, для которых выполнены оба условия:
1. в строке только одно число повторяется дважды (ровно 2 раза), остальные числа не повторяются;

2. сумма максимального и минимального среди неповторяющихся чисел строки не больше суммы повторяющихся чисел.

Ответ: 1159

```python
# инициализируем счётчик
cnt = 0

# построчно считываем файл
for s in open('9-26.csv'):
  # формируем список из строки
  row = [int(item) for item in s.split(';')] 

  # пропускаем строки, где не ровно один повторяющийся два раза элемент 
  if len(row) != len(set(row)) + 1:
    continue

  # находим сумму повторяющихся элементов  
  sum_duplicates = ( sum(row) - sum(set(row)) ) * 2
  # инициализируем минимальный и максимальный неповторяющиеся
  max_unique = -float('inf')
  min_unique = float('inf')

  # обходим поэлементно строку
  for item in row:
    if row.count(item) == 1:
      # находим максимальное неповторяющееся число
      if item > max_unique: 
        max_unique = item
      # находим минимальное неповторяющееся число
      if item < min_unique:
        min_unique = item
  
  # если сумма максимального и минимального неповторяющихся не больше суммы повторяющихся
  if min_unique + max_unique <= sum_duplicates:
    # увеличиваем счётчик
    cnt += 1

# печатаем ответ
print(cnt)


```

**10.** В файле электронной таблицы в каждой строке содержатся шесть натуральных чисел. Определите количество строк таблицы, для которых выполнены оба условия:
1. в строке только одно число повторяется трижды (ровно 3 раза), остальные числа не повторяются;
2. утроенная сумма неповторяющихся чисел строки не больше произведения повторяющихся чисел.

Ответ: 135

```python
cnt = 0

for s in open('9-26.csv'):
  row = [int(item) for item in s.split(';')] 

  # отбрасываем строки, где нет одного повторяющегося трижды или двух повторяющихся дважды чисел 
  if len(row) != len(set(row)) + 2:
    continue
  
  ok = False
  # ищем в строке число повторяющееся трижды 
  for item in row:
    if row.count(item) == 3:
      ok = True
      break

  # отбрасываем строки, где нет трёх повторяющихся чисел
  if not ok:
    continue

  sum_unique = 0
  mul_duplicates = 1
  # поэлементно обходим строку
  for item in row:
    if row.count(item) == 3:
      # находим произведение повторяющихся чисел
      mul_duplicates = item ** 3
    if row.count(item) == 1:
      # находим сумму неповторяющихся чисел
      sum_unique += item
  
  # если сумма меньше произведения
  if sum_unique <= mul_duplicates:
    # увеличиваем счётчик
    cnt += 1

# печатаем ответ
print(cnt)
```
