**Пример 1. (Чётность камней в куче).**

 Два игрока, Петя и Ваня, играют в следующую игру. Перед игроками лежит куча камней. Игроки ходят по очереди, первый ход делает Петя. За один ход игрок может добавить в кучу один камень, добавить в кучу 2 камня или увеличить количество камней в куче в три раза. При этом после каждого хода в куче должно быть чётное количество камней. Например, пусть в куче было 8 камней. Тогда за один ход можно получить кучу из 10 камней или из 24 камней (увеличить количество на 1 и получить кучу из 9 камней нельзя, т.к. после этого хода получится нечётное количество камней). Для того чтобы делать ходы, у каждого игрока есть неограниченное количество камней. Выигрывает тот игрок, после хода которого количество камней в куче становится не менее 88.

В начальный момент в куче было S камней; 1 ≤ S ≤ 87. Будем говорить, что игрок имеет выигрышную стратегию, если он может выиграть при любых ходах противника.
Ответьте на следующие вопросы:

  ***Вопрос 1.*** Известно, что Ваня выиграл своим первым ходом после первого хода Пети. Назовите минимальное значение S, при котором это возможно.

  ***Вопрос 2.*** Найдите два значения S, при которых у Пети есть выигрышная стратегия, причём Петя не может выиграть первым ходом, но может выиграть своим вторым ходом независимо от того, как будет ходить Ваня. Найденные значения запишите в ответе в порядке возрастания.

  ***Вопрос 3.*** Сколько существует значений S, при которых у Вани есть выигрышная стратегия, позволяющая ему выиграть вторым ходом при любой игре Пети.

Ответ

1) 10

2) 26 27

3) 2

```python
# кэшируем рекурсивную функцию
from functools import lru_cache
@lru_cache(maxsize=None)
def f(s):
  # условие завершения игры
  if s > 87:
    return 0
  # список оценок позиций, в которые можно сходить из текущей позиции
  moves = []
  # если при добавлении в кучу одного камня - число камней в куче станет чётным
  if (s+1) % 2 == 0:
    # разрешается добавить один камень
    moves += [f(s+1)]
  # если при добавлении в кучу двух камней - число камней в куче станет чётным
  if (s+2) % 2 == 0:
    # разрешается добавить два камя
    moves += [f(s+2)]
  # если при троекратном увеличении камней в куче - число камней в куче станет чётным
  if (s*3) % 2 == 0:
    # разрешается увеличить число камней в куче в 3 раза
    moves += [f(s*3)]
  
  # список проигранных для оппонента позиций, в которые можно попасть из текущей позиции
  loses = [x for x in moves if x <= 0]
  # если есть возможность сходить хотя бы в 1 проигрышную позицию
  if loses:
    # текущая позиция - выигрышная за -max(loses) + 1 ходов
    return 1 - max(loses)
  # если же все ходы из текущей позиции ведут в выигрышные позиции
  else:
    # текущая позиция - проигрышная за -max(moves) ходов
    return - max(moves)

# Вопрос 1
for s in range(1,88):
  # Петя создал для Вани выигрышную в 1 ход позицию
  # он мог сделать это добавив 1 или 2 камня в кучу, либо увеличив количество камней в куче в 3 раза
  # каждую ход он мог сделать только если после его завершения в куче оказывалось чётное число камней
  # найдём минимальное s, при котором это возможно
  if (s+1) % 2 == 0 and f(s+1) == 1 or (s+2) % 2 == 0 and f(s+2) == 1 or (s*3) % 2 == 0 and f(s*3) == 1:
    print(s)
    # т.к. мы ищем минимальное s - выходим из цикла при нахождение первого подходящего значения s
    break

# Вопрос 2
# заводим список для хранения искомых позиций
wins = []
for s in range(1,88):
  # f(s) = 2 - позиции, в которых Петя выигрывает своим вторым ходом 
  if f(s) == 2:
    # добавляем найденные значения s в список
    wins += [s]
# упорядочиваем найденные значения по возрастанию
print(wins[0], wins[1])


# Вопрос 3
wins = []
for s in range(1,88):
  # f(s) = -2 - позиции, в которых Ваня выигрывает своим вторым ходом 
  if f(s) == -2:
    wins += [s]
# выводим количество позиций, в которых Ваня выигрывает своим вторым ходом
print(len(wins))
```

Например, если в начале игры в куче 4 камня, Петя может первым ходом получить кучу из 5, 7 или 8 камней. Если Петя добавил 1 камень и получил кучу из 5 камней, то следующим ходом Ваня может либо добавить 3 камня (и получить 8 камней), либо удвоить количество камней в куче (их станет 10). Получить 6 камней Ваня не может, так как для этого нужно добавить 1 камень, а такой ход только что сделал Петя.
Чтобы делать ходы, у каждого игрока есть неограниченное количество камней.
Игра завершается, когда количество камней в куче становится не менее 144. Победителем считается игрок, сделавший последний ход, то есть первым получивший кучу, в которой будет 144 или больше камней. В начальный момент в куче было S камней, 1 ≤ S ≤ 143.

  ***Вопрос 1.*** Укажите такое значение S, при котором Петя не может выиграть за один ход, но при любом ходе Пети Ваня может выиграть своим первым ходом.

− Петя не может выиграть за один ход;

− Петя может выиграть своим вторым ходом независимо от того, как будет ходить Ваня.

Найденные значения запишите в ответе через пробел в порядке возрастания.
  
  ***Вопрос 3.*** Найдите значение S, при котором одновременно выполняются два условия:

– у Вани есть выигрышная стратегия, позволяющая ему выиграть первым или вторым ходом при любой игре Пети;

– у Вани нет стратегии, которая позволит ему гарантированно выиграть первым ходом.

```python
# кэшируем рекурсивную функцию
from functools import lru_cache
@lru_cache(maxsize=None)
# функция принимает дополнительный параметр type - id хода, сделанного на предыдущем шаге
def f(type, s):
  # условие завершения игры
  if s > 143:
    return 0
  # список оценок позиций, в которые можно сходить из текущей позиции
  moves = []
  # если текущий ход первый в игре
  if type == '0':
    # допустимы все возможные действия
    moves = [f('1', s+1), f('2', s*2), f('3', s+3)]
  # если на предыдущем ходе в кучу добавили 1 камень
  if type == '1':
    # допустимы все действия, кроме добавления в кучу одного камня
    moves += [f('2', s*2), f('3', s+3)]
  # если на предыдущем ходе в кучу добавили 3 камня
  if type == '3':
    # допустимы все действия, кроме добавления в кучу трёх камней
    moves += [f('1', s+1), f('2', s*2)]
  # если на предыдущем ходе количество камней в куче удволили
  if type == '2':
    # допустимы все действия, кроме удвоения количества камней в куче
    moves += [f('1', s+1), f('3', s+3)]
  
  # список проигранных для оппонента позиций, в которые можно попасть из текущей позиции
  loses = [x for x in moves if x <= 0]
  # если есть возможность сходить хотя бы в 1 проигрышную позицию
  if loses:
    # текущая позиция - выигрышная за -max(loses) + 1 ходов
    return 1 - max(loses)
  # если же все ходы из текущей позиции ведут в выигрышные позиции
  else:
    # текущая позиция - проигрышная за -max(moves) ходов
    return - max(moves)

# Вопрос 1
for s in range(1,144):
  # f('0', s) = -1 - позиции, в которых Ваня выигрывает своим первым ходом 
  if f('0', s) == -1:
    print(s)

# Вопрос 2
# заводим список для хранения искомых позиций
wins = []
for s in range(1,144):
  # f(s) = 2 - позиции, в которых Петя выигрывает своим вторым ходом 
  if f('0', s) == 2:
    # добавляем найденные значения s в список
    wins += [s]
# выводим минимальное и максимальное из найденных значений s
print(min(wins), max(wins))

# Вопрос 3
wins = []
for s in range(1,144):
  # f(s) = 2 - позиции, в которых Ваня выигрывает своим вторым ходом 
  if f('0', s) == -2:
    print(s)
```

**Пример 3. (Пропуск хода).** 

Два игрока, Петя и Ваня, играют в следующую игру. Перед игроками лежит куча камней. Игроки ходят по очереди, первый ход делает Петя. За один ход игрок может увеличить количество камней в куче в два раза или добавить в кучу один камень. Так же за всю игру можно только один раз пропустить ход - сделать ход после которого количество камней в куче не изменится, а очередь хода перейдёт к сопернику. То есть пропуск хода может единственный раз за игру сделать либо Ваня, либо Петя. Для того чтобы делать ходы, у каждого игрока есть неограниченное количество камней. Выигрывает тот игрок, после хода которого количество камней в куче становится не менее 42.

В начальный момент в куче было S камней; 1 ≤ S ≤ 41. Будем говорить, что игрок имеет выигрышную стратегию, если он может выиграть при любых ходах противника.

Ответьте на следующие вопросы:

  ***Вопрос 1.*** Известно, что Ваня выиграл своим первым ходом после неудачного первого хода Пети. Назовите минимальное значение S, при котором это возможно.
  
  ***Вопрос 2.*** Найдите значение S, при котором у Пети есть выигрышная стратегия, причём Петя не может выиграть первым ходом, но может выиграть своим вторым ходом независимо от того, как будет ходить Ваня.
  
  ***Вопрос 3.*** Найдите наименьшее и наибольшее значения S, при которых у Вани есть выигрышная стратегия. Найденные значения запишите в ответе через пробел в порядке возрастания.

Ответ

1) 11

2) 20

3) 4 19

```python
# кэшируем рекурсивную функцию
from functools import lru_cache
@lru_cache(maxsize=None)
# функция принимает дополнительный параметр sp - индикатор пропущенного хода
# если sp = 0 - ход ещё никто не пропускал
# если sp = 1 - пропуск хода уже был использован 
def f(sp, s):
  # условие завершения игры
  if s > 41:
    return 0
  # список оценок позиций, в которые можно сходить из текущей позиции
  moves = []
  # если никто ещё не воспользовался возможностью пропустить ход
  if sp == 0:
    # возможны все действия, включая пропуск хода
    # в случае использования пропуска хода sp меняет значение на 1
    moves = [f(0, s+1), f(0, s*2), f(1, s)]
  # если же к текущему моменту один из игроков уже использовал пропуск хода 
  else:
    # возможны все ходы кроме пропуска хода
    moves = [f(1, s+1), f(1, s*2)]
  # список проигранных для оппонента позиций, в которые можно попасть из текущей позиции
  loses = [x for x in moves if x <= 0]
  # если есть возможность сходить хотя бы в 1 проигрышную позицию
  if loses:
    # текущая позиция - выигрышная за -max(loses) + 1 ходов
    return 1 - max(loses)
  # если же все ходы из текущей позиции ведут в выигрышные позиции
  else:
    # текущая позиция - проигрышная за -max(moves) ходов
    return - max(moves)

# Вопрос 1
for s in range(1,42):
  # Петя создал для Вани выигрышную в 1 ход позицию
  # он сделал это, добавив в кучу 1 камень, увеличив кучу в 2 раза, или пропустив ход
  # найдём минимальное s, при котором это возможно
  if f(0, s+1) == 1 or f(0, s*2) == 1 or f(1, s) == 1:
    print(s)
    # т.к. мы ищем минимальное значение s, при нахождении первого s - выйдем из цикла 
    break

# Вопрос 2
for s in range(1,42):
   # f(s) = 2 - позиции, в которых Петя выигрывает своим вторым ходом
  if f(0, s) == 2:
    print(s)

# Вопрос 3
# заводим список для хранения искомых позиций
wins = []
for s in range(1,42):
  # f(s) < 0 - позиции, в которых у Вани есть выигрышная стратегия
  if f(0, s) < 0:
    # добавляем найденные значения s в список
    wins += [s]
# выводим минимальное и максимальное из найденных значений s
print(min(wins), max(wins))
```

**Пример 4. (неизвестное количество камней в обеих кучах).**

Два игрока, Петя и Ваня, играют в следующую игру. Перед игроками лежит две кучи камней. Игроки ходят по очереди, первый ход делает Петя. За один ход игрок может убрать из кучи один камень или уменьшить количество камней в любой куче в два раза (если количество камней нечётно, то остаётся на один камень меньше, чем убирается). Игра завершается в тот момент, когда общее количество камней в двух кучах становится не более 32, побеждает игрок, сделавший последний ход. В начальный момент в первой куче было K≥1 камней, а во второй – S≥1 камней, S+K ≥ 33.

Ответьте на следующие вопросы:

  ***Вопрос 1.*** Известно, что из начальной позиции (N; N) Ваня выигрывает первым ходом при любой игре Пети. При каком значении N это возможно?

  ***Вопрос 2.*** При K = 23, найдите минимальное и максимальное значение S, при котором у Пети есть выигрышная стратегия, причём одновременно выполняются два условия:

− Петя не может выиграть за один ход;

− Петя может выиграть своим вторым ходом независимо от того, как будет ходить Ваня.

Найденные значения запишите в ответе через пробел в порядке возрастания.

 ***Вопрос 3.*** При каком максимальном значении K для начальной пары (K; S) одновременно выполняются два условия:

– у Вани есть выигрышная стратегия, позволяющая ему выиграть первым или вторым ходом при любой игре Пети;

– у Вани нет стратегии, которая позволит ему гарантированно выиграть первым ходом.

Ответ

1) 22

2) 22 44

3) 67

```python
# кэшируем рекурсивную функцию
from functools import lru_cache
@lru_cache(maxsize=None)
def f(a, b):
  # условие завершения игры
  if a + b <= 32:
    return 0
  # список оценок позиций, в которые можно сходить из текущей позиции
  moves = [f(a-1, b), f(a, b-1)]
  # если в первой куче есть хотя бы 1 камень (страховка от зацикливания в нуле)
  if a > 0:
    # можно убрать половину камней из первой кучи 
    moves +=  [f(a//2, b)]
  # если во второй куче есть хотя бы 1 камень (страховка от зацикливания в нуле) 
  if b > 0: 
    # можно убрать половину камней из второй кучи
    moves +=  [f(a, b//2)]
  # список проигранных для оппонента позиций, в которые можно попасть из текущей позиции
  loses = [x for x in moves if x <= 0]
  # если есть возможность сходить хотя бы в 1 проигрышную позицию
  if loses:
    # текущая позиция - выигрышная за -max(loses) + 1 ходов
    return 1 - max(loses)
  # если же все ходы из текущей позиции ведут в выигрышные позиции
  else:
    # текущая позиция - проигрышная за -max(moves) ходов
    return - max(moves)

# Вопрос 1
# начинаем поиск с n = 17, потому что при n = 16 в кучах всего 32 камня и игра не состоится
for n in range(17,100):
  # f(n,n) = -1 - позиции, в которых Ваня выигрывает своим первым ходом
  if f(n, n) == -1:
    print(n)
print('---------')

# Вопрос 2
# заводим список для хранения искомых количеств камней во второй куче 
wins = []
# начинаем поиск с s = 10, чтобы в кучах на старте было хотя бы 33 (10+23) камня
for s in range(10,100):
  # f(23,s) = 2 - позиции, в которых Петя выигрывает своим вторым ходом
  if f(23,s) == 2:
    # добавляем найденные значения s в список
    wins += [s]
# выводим минимальное и максимальное из найденных значений s
print(min(wins), max(wins))
print('---------')

# Вопрос 3
wins = []
for k in range(1,100):
  for s in range(1,100):
    # f(k,s) = 2 - позиции, в которых Ваня выигрывает своим вторым ходом
    if f(k,s) == -2:
      wins += [k]
print(max(wins))
```
