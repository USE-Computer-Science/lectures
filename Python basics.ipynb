### **Циклы for**

```python
# оператор цикла for i in range обеспечивает повторное выполнение инструкций в своём теле до тех пор 
# пока значение переменной i не выходит за пределы диапазона range
for i in range(1, 101):
  print(i)
```

```python
# оператор цикла for i in string обеспечивает повторное выполнение инструкций в своём теле до тех пор 
# пока значение переменной i не примет значение последнего символа в строке string
for i in "Hello":
  print(i)
```

image.png

```python
# напечатать все числа от 1 до 10, кроме 7
for i in range(1, 11):
  if i == 7:
    continue
  print(i)
```

```python
# печатать все числа, начиная с 1 до тех пор, пока не встретится число 8
for i in range(1, 1000):
  print(i)
  if (i == 8):
    break
```

### **Функции**

image.png

- Функция оъявляется с помощью ключевого слова **def**
- Параметры (аргументы) функции передаются в скобках
- Количество параметров (аргументов) определяется разработчиком
- в конце строки объявления ставится **:**
- функция возвращает значение с помощью оператора **return**
- функция может не возвращать значение (называется процедура)
- функция может возвращать несколько значений **return a, b**

```python
# Процедура, выводящая на экран строку, принимаемую в качестве аргумента 
def my_print(s):
  print(s)
  return 
```

```python
# Вызов процедуры
my_print('Hello World!')
```

```python
# Функция принимающая 3 аргумента, и возвращающая их сумму
def sum(a, b, c):
  s = a + b + c
  return s
```

```python
# Вызов функции
answer = sum(1, 3, 5)
print(answer)
```

```python
# Функция принимающая 4 аргумента, возвращающая их произведение и признак чётности произведения
def mul(a, b, c, d):
  m = a * b * c * d
  
  is_even = 'odd'
  if m % 2 == 1:
    is_even = 'odd'
  else:
    is_even = 'even'
  
  #is_even = 'odd' if m % 2 == 1 else 'even' 

  return m, is_even
```

```python
# Вызов функции
answer, tag  = mul(1, 3, 5, 7)
print(f'answer = {answer}, this is {tag} number')
```
