#### **20. Задачи с двумя кучами камней**

**Пример 1.** 

Два игрока, Петя и Ваня, играют в следующую игру. Перед игроками лежат две кучи камней. Игроки ходят по очереди, первый ход делает Петя. За один ход игрок может добавить в одну из куч (по своему выбору) два камня или увеличить количество камней в куче в три раза. Для того чтобы делать ходы, у каждого игрока есть неограниченное количество камней. Игра завершается в тот момент, когда суммарное количество камней в кучах становится не менее 234. Победителем считается игрок, сделавший последний ход, т.е. первым получивший такую позицию, при которой в кучах будет 234 или больше камней.
В начальный момент в первой куче было 34 камня, во второй куче – S камней; 1 ≤ S ≤ 199.
Ответьте на следующие вопросы:

  ***Вопрос 1***. Известно, что Ваня выиграл своим первым ходом после неудачного первого хода Пети. Укажите максимальное значение S, когда такая ситуация возможна.

  ***Вопрос 2***. Найдите значения S, при которых у Пети есть выигрышная стратегия, причём одновременно выполняются два условия:
− Петя не может выиграть за один ход;
− Петя может выиграть своим вторым ходом независимо от того, как будет ходить Ваня.
Найденные значения запишите в ответе через пробел в порядке убывания.

  ***Вопрос 3***. Найдите такое значение S, при котором одновременно выполняются два условия:
– у Вани есть выигрышная стратегия, позволяющая ему выиграть первым или вторым ходом при любой игре Пети;
– у Вани нет стратегии, которая позволит ему гарантированно выиграть первым ходом.


***Ответ:***

1) 197

2) 64 22

3) 65

```python
# кэшируем рекурсивную функцию
from functools import lru_cache
@lru_cache(maxsize=None)
def f(a,b):
  # условие завершения игры
  if a + b >= 234:
    return 0
  # список оценок позиций, в которые можно сходить из текущей позиции
  moves = [f(a+2,b), f(a,b+2), f(a*3, b), f(a,b*3)]
  # список проигранных для оппонента позиций, в которые можно попасть из текущей позиции
  loses = [x for x in moves if x <= 0]
  # если есть возможность сходить хотя бы в 1 проигрышную позицию
  if loses:
    # текущая позиция - выигрышная за -max(loses) + 1 ходов
    return 1 - max(loses)
  # если же все ходы из текущей позиции ведут в выигрышные позиции
  else:
    # текущая позиция - проигрышная за -max(moves) ходов
    return - max(moves)

# Вопрос 1
# начинаем поиск с конца, т.к. требуется найти максимальное значение s 
for s in range(199, 1, -1):
  # Петя создал для Вани выигрышную в 1 ход позицию
  # он мог сделать это либо добавив 2 камня в одну из куч, либо увеличив количество камней в одной из куч в 3 раза
  # найдём максимальное s, при котором это возможно
  if f(34+2,s) == 1 or f(34,s+2) == 1 or f(34*3,s) == 1 or f(34,s*3) == 1:
    print(s)
    # т.к. нас интересует максимальное s - можем завершить поиск после нахождения первого подходящего s
    break

# Вопрос 2
# заводим список для хранения искомых позиций
wins = []
for s in range(1,200):
  # f(34, s) = 2  - позиции, в которых Петя выигрывает своим вторым ходом
  if f(34,s) == 2:
    # добавляем найденные значения s в список
    wins +=[s]
# упорядочиваем найденные значения s по убыванию
print(max(wins), min(wins))

# Вопрос 3
for s in range(1,200):
  # f(34, s) = -2  - позиции, в которых Ваня выигрывает своим вторым ходом
  if f(34,s) == -2:
    print(s)
```

**Пример 2.** 

Два игрока, Паша и Влад, играют в следующую игру. Перед игроками лежат две кучи камней. Игроки ходят по очереди, первый ход делает Паша. За один ход игрок может добавить в одну из куч (по своему выбору) два камня или увеличить количество камней в куче в два раза. Для того чтобы делать ходы, у каждого игрока есть неограниченное количество камней.
Игра завершается в тот момент, когда суммарное количество камней в кучах становится не менее 70. Если при этом в куче оказалось не более 85 камней, то победителем считается игрок, сделавший последний ход. В противном случае победителем становится его противник, при этом считается, что противник сделал ход.

В начальный момент в первой куче было 10 камней, во второй куче – S камней; 1 ≤ S ≤ 59.

Ответьте на следующие вопросы:
  
  ***Вопрос 1.*** Укажите значение S, при котором Паша не может победить своим первым ходом, но Влад побеждает своим первым ходом при любой игре Паши.
  
  ***Вопрос 2.*** Укажите, сколько существует значений S, при которых у Паши есть выигрышная стратегия, причём одновременно выполняются два условия:

– Паша не может гарантированно выиграть за один ход;

– Паша может выиграть своим вторым ходом независимо от того, как будет ходить Влад.

  ***Вопрос 3.*** Укажите минимальное и максимальное значения S, при которых одновременно выполняются два условия:

– у Влада есть выигрышная стратегия, позволяющая ему выиграть первым или вторым ходом при любой игре Паши;

– у Влада нет стратегии, которая позволит ему гарантированно выиграть первым ходом.

Найденные значения запишите в ответе через пробел в порядке возрастания.

Ответ

1) 29

2) 7

3) 26 43

```python
# кэшируем рекурсивную функцию
from functools import lru_cache
@lru_cache(maxsize=None)
def f(a, b):
  # условие победного завершения игры
  if a + b >= 70 and a + b <= 85:
    return 0
  # условие проигрыша
  # мы хотим, чтобы игрок сходивший в уже проигранную позицию имел позицию проигрышную в 1 ход
  # следовательно проигранная позиция оценивается как выигранная в 1 ход (второй игрок сходил из неё и выиграл)
  if a + b > 85:
    return 1
  # список оценок позиций, в которые можно сходить из текущей позиции
  moves = [f(a*2, b), f(a, b*2), f(a+2, b), f(a, b+2)]
  # список проигранных для оппонента позиций, в которые можно попасть из текущей позиции
  loses = [x for x in moves if x <= 0]
  # если есть возможность сходить хотя бы в 1 проигрышную позицию
  if loses:
    # текущая позиция - выигрышная за -max(loses) + 1 ходов
    return 1 - max(loses)
  # если же все ходы из текущей позиции ведут в выигрышные позиции
  else:
    # текущая позиция - проигрышная за -max(moves) ходов
    return - max(moves)

# Вопрос 1
for s in range(1,60):
  # f(10, s) = -1  - позиции, в которых Влад выигрывает своим первым ходом
  if f(10,s) == -1:
    print(s)

# Вопрос 2
# заводим список для хранения искомых позиций
wins = []
for s in range(1,60):
  # f(10, s) = 2  - позиции, в которых Петя выигрывает своим вторым ходом
  if f(10, s) == 2:
    # добавляем найденные значения s в список
    wins += [s]
# выводим количество выигрышных в 2 хода позиций для Пети
print(len(wins))

# Вопрос 3
wins = []
for s in range(1,60):
  # f(10, s) = 2  - позиции, в которых Влад выигрывает своим вторым ходом
  if f(10, s) == -2:
    wins += [s]
# упорядочиваем найденные значения s по возрастанию
print(min(wins), max(wins))
```

**Пример 3.** 

Два игрока, Петя и Вова, играют в следующую игру. Перед игроками лежат две кучи камней. Игроки ходят по очереди, первый ход делает Петя. За один ход игрок может убрать из одной из куч один камень или уменьшить количество камней в куче в два раза (если количество камней в куче нечётно, остаётся на 1 камень меньше, чем убирается). Предположим, например, что в первой куче было 5, а во второй 8 камней. За один ход из позиции (5, 8) можно получить любую из четырёх позиций: (4, 8), (5, 7), (2, 8), (5, 4). Игра завершается в тот момент, когда суммарное количество камней в кучах становится не более 30. Победителем считается игрок, сделавший последний ход, то есть первым получивший позицию, в которой в кучах будет 30 или меньше камней. В начальный момент в первой куче было 20 камней, во второй куче – S камней, S > 10. 

Ответьте на следующие вопросы:

***Вопрос 1.*** Найдите значение S, при котором Вова выигрывает своим первым ходом при любой игре Пети.

***Вопрос 2.*** Найдите минимальное и максимальное значение S, при котором у Пети есть выигрышная стратегия, причём одновременно выполняются два условия: 

− Петя не может выиграть за один ход; 

− Петя может выиграть своим вторым ходом независимо от того, как будет ходить Вова. 

Найденные значения запишите в ответе через пробел в порядке возрастания.

***Вопрос 3.*** Найдите значение S, при котором одновременно выполняются два условия: 

– у Вовы есть выигрышная стратегия, позволяющая ему выиграть первым или вторым ходом при любой игре Пети; 

– у Вовы нет стратегии, которая позволит ему гарантированно выиграть первым ходом.

Ответ:

1) 22

2) 23 45

3) 25

```python
# кэшируем рекурсивную функцию
from functools import lru_cache
@lru_cache(maxsize=None)
def f(a, b):
  
  # условие завершения игры
  if a + b <= 30:
    return 0
  
  # список оценок позиций, в которые можно сходить из текущей позиции
  moves = []
  # если в первой куче есть хотя бы 1 камень (страховка от ухода в минус)
  if a > 0:
    # можно убрать этот камень из первой кучи
    moves += [f(a-1, b)]
  # если во второй куче есть хотя бы 1 камень (страховка от ухода в минус)
  if b > 0:
    # можно убрать этот камень из второй кучи
    moves += [f(a, b-1)]

  # если в первой куче есть хотя бы 1 камень (страховка от зацикливания в нуле)
  if a > 0:
    # если текущее количество камней в первой куче чётно
    if a % 2 == 0:
      # можно убрать половину камней из первой кучи
      moves += [f(a//2, b)]
    # если текущее количество камней в первой куче нечётно
    else:
      # можно убрать на 1 камень  меньше, чем останется в первой куче
      moves += [ f((a//2), b)]
  # если во второй куче есть хотя бы 1 камень (страховка от зацикливания в нуле)   
  if b > 0:
    # если текущее количество камней во второй куче чётно
    if b % 2 == 0:
      # можно убрать половину камней из второй кучи
      moves += [f(a, b//2)]
    # если текущее количество камней во второй куче нечётно
    else:
      # можно убрать на 1 камень  меньше, чем останется во второй куче
      moves += [f(a, (b//2))]
  
  # список проигранных для оппонента позиций, в которые можно попасть из текущей позиции
  loses = [x for x in moves if x <= 0]
  # если есть возможность сходить хотя бы в 1 проигрышную позицию
  if loses:
    # текущая позиция - выигрышная за -max(loses) + 1 ходов
    return 1 - max(loses)
  # если же все ходы из текущей позиции ведут в выигрышные позиции
  else:
    # текущая позиция - проигрышная за -max(moves) ходов
    return - max(moves)

# Вопрос 1
for s in range(11,100):
  # f(20, s) = -1 - позиции, в которых Вова выигрывает своим первым ходом 
  if f(20, s) == -1:
    print(s)

# Вопрос 2
# заводим список для хранения искомого количества камней во второй куче
wins = []
for s in range(11,100):
  # f(20, s) = 2 - позиции, в которых Петя выигрывает своим вторым ходом 
  if f(20, s) == 2:
    # добавляем найденные значения s в список
    wins += [s]
# выводим минимальное и максимальное найденные значения s
print(min(wins), max(wins))

# Вопрос 3
for s in range(11,100):
  # f(20, s) = -2 - позиции, в которых Вова выигрывает своим вторым ходом 
  if f(20, s) == -2:
    print(s)
```
