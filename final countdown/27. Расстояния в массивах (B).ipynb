**27.1. Количество пар**

По каналу связи передаётся последовательность положительных целых чисел, все числа не превышают 1000. Рассматриваются все пары элементов последовательности, находящихся на расстоянии не меньше чем k (разница в индексах элементов пары должна быть k или более, порядок элементов в паре неважен), разность которых нечётна и, по крайней мере, один из элементов делится на 26. Необходимо найти количество таких пар.

**Входные данные**

Даны два входных файла ([файл A](https://kompege.ru/files/rLwo2mmjz.txt) и [файл B](https://kompege.ru/files/r_Z94UAtg.txt)), каждый из которых содержит в первой строке натуральное число k — расстояние между элементами, а во второй - количество чисел N (1 ≤ N ≤ 10 000 000, N > K). Каждая из следующих N строк содержит одно натуральное число, не превышающее 1000.

В ответе укажите два числа: сначала значение искомой суммы для файла А, затем – для файла B.

**Пример входных данных:**

2

6

52

6

3

26

9

102

Подходящие пары чисел это (52, 3), (52, 9).

Ответ для примера: 2.

**Ответ:** 15831 172216572903

```python
f = open('27-1A.txt')
k = int(f.readline())
n = int(f.readline())

nums = [int(x) for x in f]
cnt = 0

for i in range(n):
  for j in range(k+i, n):
    if nums[i] % 26 == 0 or nums[j] % 26 == 0:
      if abs(nums[i] - nums[j]) % 2 == 1:
        cnt += 1
        mx = max(mx, abs(nums[i] + nums[j]))

cnt
```

```python
f = open('27-1B.txt')
k = int(f.readline())
n = int(f.readline())

a = [int(x) for x in f]
cnt = 0

# счётчик чётных чисел, кратных 26
k_even_26 = 0
# счётчик чётных чисел, не кратных 26
k_even_not26 = 0
# счётчик нечётных чисел, кратных 26
k_odd_26 = 0
# счётчик нечётных чисел, кратных 26
k_odd_not26 = 0

# обходим все пары на расстоянии ровно в k элементов друг от друга
for i in range(k, n):
  # увеличиваем соответствующий счётчик для первого числа в паре
  if a[i-k] % 2 == 0 and a[i-k] % 26 == 0: k_even_26 += 1
  if a[i-k] % 2 == 0 and a[i-k] % 26 != 0: k_even_not26 += 1
  if a[i-k] % 2 != 0 and a[i-k] % 26 == 0: k_odd_26 += 1
  if a[i-k] % 2 != 0 and a[i-k] % 26 != 0: k_odd_not26 += 1

  # для чётных кратных 26 вторых чисел в паре
  if a[i] % 2 == 0 and a[i] % 26 == 0:
    # годятся все нечётные не кратные 26 первые числа
    # и  все нечётные кратные 26 первые числа
    cnt += k_odd_not26 + k_odd_26
  # для чётных не кратных 26 вторых чисел в паре
  if a[i] % 2 == 0 and a[i] % 26 != 0:
    # годятся только нечётные кратные 26 первые числа
    cnt += k_odd_26
  # и т.д.
  if a[i] % 2 != 0 and a[i] % 26 == 0: cnt += k_even_not26 + k_even_26
  if a[i] % 2 != 0 and a[i] % 26 != 0: cnt += k_even_26

# общее количество пар
print(cnt)
```

**27.2. Количество пар**

На вход программы поступает последовательность из N натуральных чисел. Рассматриваются все пары различных элементов последовательности (элементы пары не обязательно должны стоять в последовательности рядом, порядок в паре неважен). Необходимо определить количество пар, для которых сумма кратна 120, ровно один из элементов пары делится на 41, а номера элементов в последовательности отличаются не менее, чем на K.

**Входные данные**

Даны два входных файла ([файл A](https://kompege.ru/files/h3Rh08H7k.txt) и [файл B](https://kompege.ru/files/w2h5Uoi1w.txt)), каждый из которых в первой строке содержит число N - количество чисел, во второй строке K – минимальная разница между номерами элементов (1 ≤ N ≤ 10 000 000, N > K). В каждой из следующих N строк записаны элементы последовательности (все числа неотрицательные, не превышающие 2 000 000).

В ответе укажите два числа: сначала значение искомой величины для файла A, затем - для файла B.

**Типовой пример организации данных во входном файле:**

**Ответ:** 72 1087816

```python
f = open('27-2B.txt')
n = int(f.readline())
k = int(f.readline())

a = [int(x) for x in f]

cnt = 0

# счётчик первых чисел в паре с остатком от деления на 120 кратных 41
k41 = [0]*120
# счётчик первых чисел в паре с остатком от деления на 120 не кратных 41
kn41 = [0]*120

# обходим все пары на расстоянии ровно в k элементов друг от друга
for i in range(k, n):
  # если первое число в паре кратно 41
  if a[i-k] % 41 == 0:
    # увеличиваем счётчик первых чисел в паре кратных 41 с соотвествующим остатком от деления на 120
    k41[a[i-k] % 120] += 1
  else:
    # увеличиваем счётчик первых чисел в паре кратных 41 с соотвествующим остатком от деления на 120
    kn41[a[i-k] % 120] += 1

  # находим нужный остаток от деления у первых чисел, чтобы сумма пары этих чисел со 2 делилась на 120
  if a[i] % 120 == 0: ost = 0
  else: ost = 120 - a[i] % 120

  # если второе число в паре кратно 41
  if a[i] % 41 == 0:
    # образовываем с ним всевозможные пары из первых чисел не кратных 41 с подходящим остатком от деления на 120
    cnt += kn41[ost]
  # если второе число в паре не кратно 41
  else:
    # образовываем с ним всевозможные пары из первых чисел кратных 41 с подходящим остатком от деления на 120
    cnt += k41[ost]

print(cnt)
```

Предупреждение: для обработки файла B не следует использовать переборный алгоритм, вычисляющий сумму для всех возможных вариантов, поскольку написанная по такому алгоритму программа будет выполняться слишком долго.

**Ответ:** 73 1146648

```python
f = open('27-2A.txt')
n = int(f.readline())
k = int(f.readline())

nums = [int(x) for x in f]
cnt = 0

for i in range(n):
  for j in range(i+1, min(i+k+1, n)):
    if (nums[i] % 37 == 0) + (nums[j] % 37 == 0) == 1:
      if abs(nums[i] - nums[j]) % 100 == 0:
        cnt += 1

cnt
```
