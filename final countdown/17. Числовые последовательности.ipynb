#### **Кортежи**

**Кортеж** — это неизменяемый список. Кортежи заключаются в круглые, а не квадратные скобки.

**Функция zip(*lst_1*, .., *lst_n*)** принимает несколько списков (*lst_1*, .., *lst_n*) в качестве аргументов. Затем она генерирует список кортежей, которые содержат соответствующие элементы из каждого списка, переданного в функцию.

Более формально:

**Функция zip()** возвращает итератор кортежей, где i-й кортеж содержит i-й элемент из каждой итерации аргументов.

#### **Примеры**

**Пример 1.** В файле содержится последовательность целых чисел. Элементы последовательности могут принимать целые значения от $–10$ $000$ до $10$ $000$ включительно. Определите количество пар последовательности, в которых сумма элементов меньше минимального положительного элемента последовательности, кратного $19$. Гарантируется что такой элемент в последовательности есть. В ответе запишите количество найденных пар, затем максимальную из сумм элементов таких пар. В данной задаче под парой подразумевается два идущих подряд элемента последовательности.

**Ответ:** 4925 36

```python
# создаём поток для чтения файла
f = open('17-1.txt')
# считываем числовую последовательность из файла в список с помощью генератора списка
nums = [int(x) for x in f]

# находим минимальный кратный 19 положительный элемент последовательности (минимум)
min19 = min([x for x in nums if abs(x) % 19 == 0 and x > 0])

# получаем список сумм пар элементов, сумма которых меньше минимума
sums = [a + b for a, b in zip(nums, nums[1:]) if a + b < min19]

# печатаем количество элементов в списке, и его максимальный элемент
print(len(sums), max(sums))
```

**Пример 2.** В файле содержится последовательность целых чисел. Элементы последовательности могут принимать целые значения от $1$ до $100$ $000$ включительно. Определите количество пар элементов последовательности, в которых остаток от деления хотя бы одного из элементов на $117$ равен минимальному элементу последовательности. В ответе запишите количество найденных пар, затем минимальный квадрат суммы элементов таких пар. В данной задаче под парой подразумевается два идущих подряд элемента последовательности.

**Ответ:** 143 67584841

```python
# создаём поток для чтения файла
f = open('17-2.txt')
# считываем числовую последовательность из файла в список с помощью генератора списка
nums = [int(x) for x in f]

# находим минимальный элемент последовательности (минимум)
mn = min(nums)

# получаем список квадратов суммa пар элементов, остаток от деления хотя одного из которых = минимуму
sums = [(a + b)**2 for a, b in zip(nums, nums[1:]) if abs(a) % 117 == mn or abs(b) % 117 == mn]

# печатаем количество элементов в списке, и его минимальный элемент
print(len(sums), min(sums))
```

**Пример 3.** В файле содержится последовательность целых чисел. Элементы последовательности могут принимать целые значения от $−10\;000$ до $10\;000$ включительно. Определите количество троек элементов последовательности, в которых произведение чётно, а сумма элементов заканчивается на $3$. В ответе запишите сначала количество найденных троек, а затем максимальный из минимальных элементов таких троек. В данной задаче под тройкой подразумевается три идущих подряд элемента последовательности.

**Ответ:** 727 8853

```python
# создаём поток для чтения файла
f = open('17-1.txt')
# считываем числовую последовательность из файла в список с помощью генератора списка
nums = [int(x) for x in f]

# получаем список минимальных элементов троек, произведение которых чётно, а сумма заканчивается на '3'
mins = [min(a,b,c) for a,b,c in zip(nums,nums[1:],nums[2:]) if abs((a*b*c)) % 2 == 0 and str(a+b+c)[-1] == '3']

# печатаем длину полученного списка и его максимальный элемент
print(len(mins), max(mins))
```
