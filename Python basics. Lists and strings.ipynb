### **Обход списка с помощью циклов**

```python
t = [1,2,3,1,1,3]
```

```python
# обход списка по индексам (позволяет добавлять и удалять элементы списка в теле цикла)
for i in range(len(t)):
  print((t[i]))
```

```python
# обход списка по элементам (только для чтения) 
# изменение списка в теле цикла вызывает неопределённое поведение программы
for item in t:
  print(item)
```

```python
# удаление элементов из списка обходом по элементам
# неопределённое поведение устраняется за счёт использования второго списка ts[]
ts =[]
for i in t:
  if i == 1:
    continue
  ts.append(i)
print(ts)
```

### **Методы списков**

image.png

```python
# создание списка
languages = ['assembler', 'fortran', 'basic', 'c']
```

```python
# добавление элемента в конец списка
languages.append('python')
print(languages) 
```

```python
# добавление элемента по индексу
languages.insert(3,'pascal')
print(languages) 
```

```python
# удаление элемента по индексу
languages.pop(4)
print(languages) 
```

```python
# удаление последнего элемента
languages.pop()
print(languages) 
```

```python
# удаление первого вхождения элемента в список по значению (самое ресурсозатратное удаление)
languages.remove('pascal')
print(languages)
```

```python
# индекс первого элемента с заданным значением
i = languages.index('fortran')
print(i)
```

```python
# количество элементов с заданным значением в списке
n = languages.count('basic')
print(n)
```

```python
# сортировка списка по возрастанию в алфавитном порядке
languages.sort()
print(languages)
```

```python
# инвертирование списка
languages.reverse()
print(languages)
```

```python
# копирование списка по ссылке
l = languages
l.append('python')
print(f'languages={languages}')
print(f'l={l}')
```

```python
# копирование списка по значению
l = languages.copy()
l.remove('python')
print(f'languages={languages}')
print(f'l={l}')
```

### **Функции для работы со списками**

image.png

```python
num_views = [9, 3, 1, 0.2, 250, 115, 150, 85, 120, 200] # Snake case
#numViews = [9, 3, 1, 0.2, 250, 115, 150, 85, 120, 200]  # Camel case 
```

```python
# возврат длины списка
m = len(num_views)
print(m)
```

```python
# возврат максимального по значению элемента списка 
m = max(num_views)
print(m)
```

```python
# возврат минимального по значению элемента списка 
m = min(num_views)
print(m)
```

```python
# возврат суммы значений элемента списка 
m = sum(num_views)
print(m)
```

```python
# сортировка списка по возрастанию в алфавитном порядке
l = sorted(num_views)
print(f'num_views = {num_views}')
print(f'l = {l}')
```

```python
# возвращает True, если условие выполняется хотя бы одного элемента списка 
m = any(i > 200 for i in num_views)
print(m)
```

```python
# возвращает True, если условие выполняется для всех элементов списка
m = all(i > 200 for i in num_views)
print(m)
```

```python
# создание списка из строки
s = 'string' 
l = list(s)
print(f's = {s}')
print(f'l = {l}')
```

### **Работа со строками**

```python
# создание строки
lang = 'Assembler'
```

```python
# возврат длины строки
l = len(lang)
print(l) 
```

```python
# добавление подстроки в конец списка
lang = lang + 'Py'
print(lang) 
```

```python
# добавление подстроки по индексу
lang = lang[:10] + 'sss' + lang[10:]
print(lang) 
```

```python
# удаление подстроки по индексам
lang = lang[:2] + lang[3:]
print(lang) 
```

```python
# удаление последнего элемента
lang = lang[:-1]
print(lang) 
```

```python
# замена заданного количества вхождений подстроки на другую подстроку replace(заменяемая подстрока, новая подстрока, количество замен)
lang = lang.replace('s','', 1)
print(lang)
```

```python
# замена всех вхождений подстроки на другую подстроку replace(заменяемая подстрока, новая подстрока)
lang = lang.replace('s','')
print(lang)
```

```python
# индекс первого вхождения подстроки find(искомая подстрока)
i = lang.find('mble')
print(i)
```

```python
# количество заданных подстрок в строке count(подсчитываемая подстрока)
n = lang.count('mb')
print(n)
```

```python
# сортировка строки по возрастанию в алфавитном порядке с помощью списка
s = sorted(lang)
lang = ''.join(s)
print(s)
print(lang)
```

```python
# инвертирование строки
lang = lang[::-1]
print(lang)
```

```python
# приведение строки к верхнему регистру
lang = lang.upper() 
print(lang)
```

```python
# приведение строки к нижнему регистру
lang = 'а1баб234АВ'
lang = lang.lower() 
print(lang)
```

```python
# форматирование строки
lang = 'Peter is {} student, he likes play {}'.format('Computer science', 'football')
print(lang)

sub = 'Mathematics'
sport = 'volleyball'
lang = 'Peter is {} student, he likes play {}'.format(sub, sport)
print(lang)
```

```python
# деление строки на несколько строк по заданному разделителю str.split('разделитель')
# str.split('разделитель', максимальное количество разделений)
l = lang.split()
print('деление строки по пробелу до конца строки: ', l)

l = lang.split(' ', 5)
print('деление строки по пробелу 5 раз: ', l)

l = lang.split(',')
print('деление строки по запятой: ', l)
```

```python
# создание строки из итерируемого объекта символов с заданным разделителем
# 'разделитель'.join(итерируемый объект)
lst = ['1','2','3','4','5']
l = ','.join(lst)
print('слияние подстрок с разделителем запятая: ', l)

l = '\n'.join(lst)
print('слияние подстрок с разделителем перевод строки:\n', l)
```

```python
# перевод десятичного числа в двоичную систему счисления
a = bin(5)[2:]
print(a)
```

```python
# перевод десятичного числа в восьмиричную систему счисления
b = oct(25)[2:]
print(b)
```

```python
# перевод десятичного числа в шестнадцатиричную систему счисления
c = hex(49)[2:]
print(c)
```

```python
# возврат в десятичную из любой системы счисления int(число, основание системы счисления)
int(b,8)
```
