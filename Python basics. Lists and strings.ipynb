```python
num_views = [9, 3, 1, 0.2, 250, 115, 150, 85, 120, 200] # Snake case
#numViews = [9, 3, 1, 0.2, 250, 115, 150, 85, 120, 200]  # Camel case 
```

```python
#   
m = len(num_views)
print(m)
```

```python
#       
m = max(num_views)
print(m)
```

```python
#       
m = min(num_views)
print(m)
```

```python
#      
m = sum(num_views)
print(m)
```

```python
#       
l = sorted(num_views)
print(f'num_views = {num_views}')
print(f'l = {l}')
```

```python
#  True,         
m = any(i > 200 for i in num_views)
print(m)
```

```python
#  True,       
m = all(i > 200 for i in num_views)
print(m)
```

```python
#     find( )
i = lang.find('mble')
print(i)
```

```python
#      count( )
n = lang.count('mb')
print(n)
```

```python
#  
lang = 'Peter is {} student, he likes play {}'.format('Computer science', 'football')
print(lang)

sub = 'Mathematics'
sport = 'volleyball'
lang = 'Peter is {} student, he likes play {}'.format(sub, sport)
print(lang)
```

```python
#         str.split('')
# str.split('',   )
l = lang.split()
print('      : ', l)

l = lang.split(' ', 5)
print('    5 : ', l)

l = lang.split(',')
print('   : ', l)
```

```python
#         
# ''.join( )
lst = ['1','2','3','4','5']
l = ','.join(lst)
print('    : ', l)

l = '\n'.join(lst)
print('     :\n', l)
```
