– все числа в строке различны;

Обозначим через ДЕЛ(n, m) утверждение «натуральное число n делится без остатка на натуральное число m»; и пусть на числовой прямой дан отрезок B = [50; 70]. Для какого наибольшего натурального числа А формула

$ДЕЛ(x, A) ∨ ((x ∈ B) → ¬ДЕЛ(x, 16))$

**15.2.**

На числовой прямой даны три отрезка: D = [15; 40], C = [21; 63] и А = [7; E]. Укажите наименьшее возможное целое значение E такое, что формула

$(x ∈ D) → ((¬(x ∈ C) ∧ ¬(x ∈ A)) → ¬(x ∈ D))$

истинна (то есть принимает значение 1 при любом значении переменной х).

**Ответ:** 20

```python
def f(x,e):
  D = 15 <= x <= 40
  C = 21 <= x <= 63
  A = 7 <= x <= e
  return D <= (((not C) and (not A)) <= (not D))

for e in range(-1000, 1000):
  if all(f(x,e) for x in range(-1000,1000)):
    print(e)
    break
```

$F(n) = 0$ , при $n = 0$,

$F(n) = F(n/2) - 1$ , при чётных $n > 0$,

$F(n) = F(n-1) + 1$ , при нечётных $n > 0$.


Сколько существует чисел $n < 1000$, для которых $F(n) = 0$?

**Ответ:** 41

Сколько существует программ, которые преобразуют исходное число 3 в число 79, и при этом траектория вычислений содержит число 11 и не содержит число 23. Также программа не должна содержать двух команд «Прибавь 1» подряд.

**Ответ:** 812266767

```python
from functools import lru_cache
@lru_cache(maxsize=None)

def f(cur, fin, t, is11):
  if cur > 79 or cur == 23:
    return 0
  if cur == fin and is11:
    return 1
  if cur == 11:
    is11 = True
  if t == '1':
    return f(cur+2, fin, '', is11) + f(cur*2, fin, '', is11)
  else:
    return f(cur+2, fin, '', is11) + f(cur*2, fin, '', is11) + f(cur+1, fin, '1', is11)

f(3,79,'', 0)
```

На автодороге расположено два центра переработки отходов, каждый в одном из пунктов сбора мусора. Расстояние между центрами переработки одинаково, независимо от направления движения по кольцевой автодороге. Определите минимальную суммарную стоимость доставки мусора из всех пунктов сбора в центры переработки отходов.

**Входные данные**

В первой строке входного файла содержится число: N (1 ≤ N ≤ 10 000 000) – количество пунктов сбора мусора на кольцевой автодороге. В каждой из следующих N строк находится число – количество мусора в контейнере (все числа натуральные, количество мусора в каждом пункте не превышает 1000). Числа указаны в порядке расположения контейнеров на автомагистрали, начиная с первого километра.

**Выходные данные**

В ответе укажите минимальную возможную суммарную стоимость доставки мусора из всех пунктов сбора в центры переработки отходов.

**Пример входного файла**

6

8

20

5

13

7

19

При таких исходных данных, если контейнеры установлены на каждом километре автодороги, необходимо открыть центры переработки в пунктах 2 и 5. В этом случае сумма транспортных затрат составит: 8 + 5 и 13 + 19 = 45.

Ответ для примера: 45

**Ответ:** 65049

```python
# создаём поток для чтения исходного файла
f = open("27-1.txt")
# считываем количество элементов последовательности
n = int(f.readline())

# считываем количество мусора в каждом пункте сбора
vols = [int(x) for x in f]

# список стоимостей доставки мусора к каждому потенциальному месту расположения завода
costs = [0] * (n//2)


# для каждого мусорного бака
for i in range(n//2):
  j = i + n//2
  # обходим все остальные баки
  for k in range(n):
    # находим расстояние от каждого контейнера k до завода i
    r1 = min(abs(i-k), n - abs(i-k))
    # находим расстояние от каждого контейнера k до завода j
    r2 = min(abs(j-k), n - abs(j-k))
    # везём туда, куда ближе
    r = min(r1,r2)
    # считаем стоимость затрат на транспортировку мусора из k до ближайшего завода
    costs[i] += r * vols[k]

# печатаем ответ
print(min(costs))
```

```python
# создаём поток для чтения исходного файла
f = open("27-2.txt")
# считываем количество элементов последовательности
a, b = map(int,f.readline().split())
n = int(f.readline())

dots = [list(map(int,s.split())) for s in f]


def belong(dot, a1,a2,b1,b2):
  return dot[0] > a1 and dot[0] < a2 and dot[1] > b1 and dot[1] < b2

mx = 0
p = 0

for a1 in range(a+1):
  for a2 in range(a1,a+1):
    for b1 in range(b+1):
      for b2 in range(b1, b+1):
        if any(belong(dot, a1,a2,b1,b2) for dot in dots):
          continue
        if (a2-a1)*(b2-b1) > mx:
          mx = (a2-a1)*(b2-b1)
          p = 2*(a2-a1) + 2*(b2-b1)
        elif (a2-a1)*(b2-b1) == mx:
          p = min(p,2*(a2-a1) + 2*(b2-b1))

p
```
