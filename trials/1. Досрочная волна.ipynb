**Задание 5.**

На вход алгоритма подаётся натуральное число $N$. Алгоритм строит по нему новое число R следующим образом.

1. Строится двоичная запись числа $N$.

2. Далее эта запись обрабатывается по следующему правилу:

  а) если число кратно 3, тогда в конец дописывается три младших разряда полученной двоичной записи,

  б) если число не кратно 3, тогда в конец дописывается двоичная последовательность, являющаяся результатом умножения 3 на остаток от деления числа $N$ на 3.

Полученная таким образом запись является двоичной записью искомого числа $R$.

Например, для исходного числа $5_{10} = 101_2$ результатом является число $101110_2 = 46_{10}$, а для исходного числа $9_{10} = 1001_2$ результатом является число $1001001_2 = 73_{10}$. Укажите наибольшее число $N$, после обработки которого с помощью этого алгоритма получается число $R$, меньшее 100. В ответе запишите это число в десятичной системе счисления.

**Ответ:** 22

```python
def R(n):
  b = bin(n)[2:]
  if n % 3 == 0:
    b = b + b[-3:]
  else:
    b = b + bin(3*(n%3))[2:]
  return int(b,2)

answer = 0
for n in range(200):
  if R(n) < 100:
    answer = n
answer
```

**Задание 6.**

Исполнитель Черепаха действует на плоскости с декартовой системой координат. В начальный момент Черепаха находится в начале координат, её голова направлена вдоль положительного направления оси ординат, хвост опущен. При опущенном хвосте Черепаха оставляет на поле след в виде линии. В каждый конкретный момент известно положение исполнителя и направление его движения. У исполнителя существует две команды:

**Вперёд n** (где n – целое число), вызывающая передвижение Черепахи на n единиц в том направлении, куда указывает её голова, и **Направо m** (где m – целое число), вызывающая изменение направления движения на m градусов по часовой стрелке. 

Запись **Повтори k [Команда1 Команда2 … КомандаS]** означает, что последовательность из S команд повторится k раз. 

<br>

Черепахе был дан для исполнения следующий алгоритм:
<br><br>
**Вправо 315**

**Повтори 7 [Вперёд 16 Направо 45 Вперёд 8 Направо 135]**
<br><br>

Определите, сколько точек с целочисленными координатами будут находиться внутри области, ограниченной линией, заданной данным алгоритмом. Точки на линии учитывать не следует. 

**Ответ:** 77

```python
from turtle import *

def belong(item, color):
    return any(i > 3 and canvas.itemcget(i, 'fill') == color for i in item)

# отключаем анимацию
tracer(0)
# убираем значёк черепахи (лишняя точка)
hideturtle()
# задаём коэффициент масштабирования
r = 50
# задаём размер области поиска пересечений
e = 0.1
# подгоняем размер экрана под размер фигуры
screensize(50 * r, 50 * r)

# задаём цвета границы (чёрный) и заливки (красный)
color('black', 'red')

left(90)
right(315)
# заливаем полученную фигуру
begin_fill()
# отрисовка фигуры
for i in range(2):
    forward(16*r)
    right(45)
    forward(8*r)
    right(135)
end_fill()
up()

cnt = 0
canvas = getcanvas()
# отрисовываем координатные точки
for x in range(-50*r, 50*r, r):
    for y in range(-50*r, 50*r, r):
        item = canvas.find_overlapping(x-e, y-e, x+e, y+e)
        if belong(item, 'red') and not belong(item,'black'):
                cnt += 1
                goto(x,-y)
                dot(5, 'blue')
print(cnt)

# обновляем холст
update()
# завершаем работу с исполнителем черепаха
done()
```

На выполнение Редактору дана следующая программа:

<pre>
НАЧАЛО
  ПОКА нашлось (25) ИЛИ нашлось (355) ИЛИ нашлось (555)
    ЕСЛИ нашлось (25) ТО
      заменить (25,3)
    КОНЕЦ ЕСЛИ

    КОНЕЦ ЕСЛИ
    ЕСЛИ нашлось (555) ТО

**Задание 19.**

Два игрока, Петя и Ваня, играют в следующую игру. Перед игроками лежит куча камней. Игроки ходят по очереди, первый ход делает Петя. За один ход игрок может добавить в кучу один или четыре камня, либо увеличить количество камней в куче в четыре раза. У каждого игрока есть неограниченное количество камней, чтобы делать ходы.

Игра завершается в тот момент, когда количество камней в куче становится не менее 78. Победителем считается игрок, сделавший последний ход, т.е. первым получивший кучу из 78 или более камней. В начальный момент в куче было S камней; 1 ≤ S ≤ 77. Будем говорить, что игрок имеет выигрышную стратегию, если он может выиграть при любых ходах противника.

Укажите минимальное значение S, при котором Ваня может выиграть своим первым ходом после любого хода Пети.

**Ответ:** 19

```python
from functools import lru_cache

@lru_cache(maxsize=None)
def f(s):
  if s >= 78:
    return 0
  moves = [f(s+1), f(s*4), f(s+4)]
  loses = [x for x in moves if x <= 0]
  if loses:
    return -max(loses) + 1
  else:
    return -max(moves)

for s in range(1,78):
  if f(s) == -1:
    print(s)
```

**Задание 20.**

Для игры, описанной в задании 19, найдите два минимальных значения S, при котором у Пети есть выигрышная стратегия вторым ходом, при этом он не может гарантировано выиграть за один ход. 

**Ответ:** 15 18

```python
from functools import lru_cache

@lru_cache(maxsize=None)
def f(s):
  if s >= 78:
    return 0
  moves = [f(s+1), f(s*4), f(s+4)]
  loses = [x for x in moves if x <= 0]
  if loses:
    return -max(loses) + 1
  else:
    return -max(moves)

for s in range(1,78):
  if f(s) == 2:
    print(s)
```

```python
from functools import lru_cache

@lru_cache(maxsize=None)
def f(s):
  if s >= 78:
    return 0
  moves = [f(s+1), f(s*4), f(s+4)]
  loses = [x for x in moves if x <= 0]
  if loses:
    return -max(loses) + 1
  else:
    return -max(moves)

for s in range(1,78):
  if f(s) == -2:
    print(s)
    break
```

**Ответ:** 586 3

```python
f = open('26.txt')
k = int(f.readline())
n = int(f.readline())
#k, n = map(int,f.readline().split())

# список багажей [время попытки сдачи, время возвращения]
bags = []
# считываем данные по заявкам на сдачу багажей
for i in range(n):
  bags += [list(map(int,f.readline().split()))]
# упорядочиваем время сдачи в хронологическом порядке
bags = sorted(bags)
# на случай сохранения очерёдности в файле
#bags = sorted(bags, key = lambda x: x[0])

# список камер хранения (в каждой ячейке хранится время её очередного освобождения)
cells = [0]*k

# функция освобождения ячеек от багажа в момент времени m
# освобождаются все ячейки, срок хранения в которых закончился на m-1 минуте 
def free(m):
  for i in range(len(cells)):
    if cells[i] < m:
      cells[i] = 0

# функция сдачи багажа в камеру хранения
# в первую по индексу свободную ячейку записывается время завершения хранения текущего багажа
def put(bag):
  for i in range(len(cells)):
    # если нашлась первая свободная ячейка 
    if cells[i] == 0:
      # записываем в неё время хранения текущего багажа
      cells[i] = bag[1]
      # возвращаем индекс этой ячейки (согласно условию, начинаются с единицы)
      return i+1
  # если ни одной свободной ячейки не оказалось - возвращаем 0
  # (багаж сдать не удалось)
  return 0  

# суммарное количество принятых на хранение багажей 
cnt = 0
# индекс ячейки, в которую удалось положить последний багаж 
last_index = 0

# обходим список багажей
for bag in bags:
  # освобождаем ячейки, которые должны быть свободны к минуте сдачи текущего багажа
  free(bag[0])
  # пробуем сдать багаж
  index = put(bag)
  # если багаж принят на хранение
  if index != 0:
    # увеличиваем счётчик сданных багажей
    cnt += 1
    # сохраняем индекс ячейки, в которую удалось положить текущий багаж 
    last_index = index
# печатаем ответ
print(cnt, last_index)
  
```

Метеорологическая станция ведёт наблюдение за количеством выпавших осадков. Показания записываются каждую минуту в течении N минут. Определяется пара измерений, между которыми прошло не менее K минут. Найдите максимальную сумму показаний среди таких пар.

**Входные данные**

Даны два входных файла (A и B), каждый из которых в первой строке содержит число N - количество измерений, во второй строке K - минимальное количество минут между искомыми измерениями. В каждой из следующих N строк находится число: количество выпавших осадков.

**Выходные данные**

В ответе укажите два числа: сначала значение искомой величины для [файла A](https://kompege.ru/files/bd4-QBiYO.txt), затем - для [файла B](https://kompege.ru/files/AnakGi6eG.txt).



**Типовой пример организации данных во входном файле**

3

При таких исходных данных ответом будет 45

**Предупреждение:** для обработки файла B не следует использовать переборный алгоритм, вычисляющий сумму для всех возможных вариантов, поскольку написанная по такому алгоритму программа будет выполняться слишком долго. 

**Ответ:** 174902 3094684

```python
f = open('27A.txt')
n = int(f.readline())
k = int(f.readline())

# список наблюдений
nums = [int(s) for s in f.readlines()]

# максимальная сумма 2 показаний, полученных с разницей не менее чем в k минут
mx = 0
# генерируем все возможные пары на расстоянии не менее k друг от друга
for i in range(0,n-k):
  for j in range(i+k,n):
    # если текущая сумма пары больше максимальной суммы, найденной до этого
    # переписываем значение максимума
    mx = max(mx, nums[j]+nums[i])

# печатаем ответ
print(mx)
```

```python
# функция эффективного поиска максимальной суммы пары на расстоянии не менее, чем k в массиве arr
def getMaxPairSum(arr, n, k):
    # массив префиксных максимумов
    preMax = [0]*n
 
    # префиксный максимум нулевого элемента списка - это сам нулевой элемент
    preMax[0] = arr[0]

    # префиксные максимумы списка [1,4,3,5,8,1,5] - это [1,4,4,5,8,8,8]
    for i in range(1, n):
        preMax[i] = max(preMax[i - 1], arr[i])
 
    # максимальная сумма 2 показаний, полученных с разницей не менее чем в k минут
    mx = -float('inf')
 
    # т.к. расстояние должно быть не меньше k - начинаем поиск c k-го элемента
    for i in range(k, n):
      # если текущая сумма пары больше максимальной суммы, найденной до этого
      # переписываем значение максимума
      mx = max(mx, arr[i] + preMax[i - k])

    return mx

f = open('27A.txt')
n = int(f.readline())
k = int(f.readline())
# список наблюдений
nums = [int(s) for s in f.readlines()]
# печатаем максимальную найденную сумму пары
print(getMaxPairSum(nums, n, k))
```
