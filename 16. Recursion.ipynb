** 1.9.**     $F(n)$,  $n$    ,   :

** 2.5.**     $F(n)$  $G(n)$,  $n$  ,   :

$F(1) = 1; G(1) = 1;$

$F(n) = 3F(n-1) - G(n-1)$,  $n \geqslant  2$

$G(n) = F(n-1) - 3G(n-1)$,  $n \geqslant  2$

   $G(1123) / F(1120)$

```python
@lru_cache
def F(n):
  if n == 1:
    return 1
  return 3 * F(n-1) - G(n-1)

@lru_cache
def G(n):
  if n == 1:
    return 1
  return F(n-1) - 3 * G(n-1)

#G(1123) / F(1120)

#    G(n+3) / F(n)   n   1  17:
print('n', 'G(n+3) / F(n)')
for n in range(1, 18):
  print(n, G(n+3) / F(n))
```
