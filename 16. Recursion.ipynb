#### **Вычисление значений рекурсивных функций**

**Пример 1.1.** Алгоритм вычисления функции $F(n)$, где $n$ – целое положительное число, задан следующими соотношениями:

$F(0) = 1$,

$F(n) = n \cdot F(n-1)$, если $n > 0$.


Чему равно значение выражения $F(7)$?

Ответ: 5040

```python
# создаём рекурсивную функцию для вычисления факториала
def F(n):
  # при n = 0
  if n == 0:
    # возвращаем 1
    return 1
  # в остальных случаях возвращаем F(n-1) * n
  return n * F(n-1)

# находим 7! = 7 * 6 * 5 * 4 * 3 * 2 * 1
print(F(7))
```

**Пример 1.2.** Алгоритм вычисления значения функции $F(n)$, где $n$ – целое неотрицательное число, задан следующими соотношениями:

$F(n) = 1$, если $n < 3$,

$F(n) = F(n – 2) - F(n – 1) + 3$, если $n > 2$ и число $n$ чётное,

$F(n) = F(n – 1) - F(n – 2) - 1$, если $n > 2$ и число $n$ нечётное.

Определите значение $F(10)$.

Ответ: 1

```python
# создаём рекурсивную функцию, опирающуюся на 2 предыдущих значения F(n-1), F(n-2)
def F(n):
  # при n = 0, 1, 2
  if n < 3:
    # F(n) = 1
    return 1
  # при чётных n > 2
  if n % 2 == 0:
    return F(n-2) - F(n-1) + 3
  # при нечётных n > 2
  return F(n-1) - F(n-2) - 1

print(F(10))
```

**Пример 1.3.** Алгоритм вычисления функции $F(n, m)$, где $n$ и $m$ – натуральные числа, задан следующими соотношениями:

$F(n, m) = 0$, если $m > n,$

$F(n, m) = 1 + F(n, m + 1)$, если $m ≤ n$ и $n$ делится на $m$,

$F(n, m) = F(n, m + 1)$, если $m ≤ n$ и $n$ не делится на $m$.

Чему равно значение выражения F(12, 2)?

Ответ: 5

```python
# рекурсивная функция с двумя параметрами для поиска всех делителей числа n, начиная с m
def F(n,m):
  if m > n:
    return 0
  if n % m == 0:
    return 1 + F(n,m+1)
  return F(n,m+1)

# у 12 пять делителей не меньших, чем число 2: [2, 3, 4, 6, 12]
F(12,2)
```

**Пример 1.4.** Алгоритмы вычисления функций $F(n)$ и $G(n)$ заданы следующими соотношениями:

$G(0) = F(0) = 1$,

$F(n) = 2G(n-1) + 2$, при $n > 0$,

$G(n) = 1 - F(n)$, при $n > 0$,

Чему равно значение $F(11)$?

Ответ: 2732

```python
# функция F зависит от функции G
def F(n):
  if n == 0:
    return 1
  return 2*G(n - 1) + 2

# в то же время функция G зависит от функции F
def G(n):
  if n == 0:
    return 1
  return 1 - F(n)

# печатаем ответ
print(F(11))
```

**Пример 1.5.** Последовательность чисел Падована задается рекуррентным соотношением:

$F(1) = 1$

$F(2) = 1$

$F(3) = 1$

$F(n) = F(n–3) + F(n–2)$, при $n > 3$, где $n$ – натуральное число.

Чему равно сотое число в последовательности Падована?

Ответ: 888855064897

```python
# создадим словарь вычисленных значений рекурсии {1: 1, 2: 1, ... 100: 888855064897}
cache = dict()

def F(n):
  # если F(n) уже закэшировано (значение F(n) для текущего n уже есть в словаре)
  if n in cache.keys():
    # возвращаем ранее вычисленное значение F(n) из словаря
    return cache[n]
  # если же в словаре ещё нет вычисленного значения F(n) для текущего n 
  if n <= 3:
    # вычисляем F(n) и добавляем его в словарь
    cache[n] = 1
    # возвращаем вычисленное значение
    return cache[n]
  if n > 3:
    # вычисляем F(n) и добавляем его в словарь
    cache[n] = F(n-3) + F(n-2)
    # возвращаем вычисленное значение
    return cache[n]

# находим сотое число в последовательности Падована
print(F(100))
```

```python
# импортируем декоратор lru_cache из модуля functools
from functools import lru_cache

# кэшируем значения рекурсивной функции с помощью декоратора 
@lru_cache()
def F(n):
  if n <= 3:
    return 1
  if n > 3:
    return F(n-3) + F(n-2)

# находим сотое число в последовательности Падована
print(F(100))
```

**Пример 1.6.** Алгоритм вычисления функции $F(n)$, где $n$ – целое положительное число, задан следующими соотношениями:

$F(0) = 1$,

$F(n) = n \cdot F(n-1)$, если $n > 0$.


Чему равна сумма цифр $F(1050)$?

Ответ: 11250

```python
import sys
# увеличиваем глубину рекурсии до 2000
sys.setrecursionlimit(2000)

# создаём рекурсивную функцию для вычисления факториала
def F(n):
  # при n = 0
  if n == 0:
    # возвращаем 1
    return 1
  # в остальных случаях возвращаем F(n-1) * n
  return n * F(n-1)


# вычисляем 1050! 
n = F(1050)
# инициализируем переменную для подсчёта суммы цифр в числе 1050!
sm = 0
# пока в числе n есть хотя бы одна значащая цифра
while n > 0:
  # прибавляем к сумме последнюю цифру (младший разряд) числа n
  sm += n % 10
  # удаляем из n младший разряд
  n = n // 10
# печатаем ответ
print(sm)

# сумму цифр можно было посчитать и с помощью преобразования 1050! к списку цифр этого числа
nums = [*map(int,str(abs(F(1050))))] # модуль необходим для корректной работы с отрицательными числами (если бы они были)
# тогда сумма цифр числа - это сумма полученного списка
print(sum(nums))
```

**Пример 1.7.** Алгоритм вычисления значения функции $F(n)$, где $n$ – целое неотрицательное число, задан следующими соотношениями:

$F(n) = 0$ при $n = 0$

$F(n) = F(n/2) - 1$ при чётных $n > 0$

$F(n) = 1 + F(n–1)$ при нечётных $n > 0$

Сколько существует чисел $n$, меньших $500$, для которых значение $F(n)$ будет равно $0$?

Ответ: 33

```python
def F(n):
  if n == 0:
    return 0
  if n % 2 == 0:
    return F(n/2) - 1
  return 1+F(n-1)  

# создадим счётчик значений параметра n, при котором рекрсивная функция F(n) равна нулю
cnt = 0
for i in range(500):
  # значение функции при текущем n равно нулю 
  if F(i) == 0:
    # увеличиваем счётчик
    cnt += 1

# печатаем ответ
print(cnt)
```
