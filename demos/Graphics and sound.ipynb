**1.** Какой минимальный объём памяти (Кб) нужно зарезервировать, чтобы можно было сохранить любое растровое изображение размером 128×128 пикселей при условии, что в изображении могут использоваться 256 различных цветов?

**Ответ:** 16

```python
import math
a = b = 128
i = math.log2(256)
V = a*b*i
print (f'Объём изображения: {V} бит')
kb = V / 1024 / 8
print (f'Необходимо зарезервировать: {kb} Кбайт')
```

**2.** Автоматическая камера производит растровые изображения размером 200×256 пикселей. Для кодирования цвета каждого пикселя используется одинаковое количество бит, коды пикселей записываются в файл один за другим без промежутков. Объём файла с изображением не может превышать 65 Кбайт без учёта размера заголовка файла. Какое максимальное количество цветов можно использовать в палитре?

**Ответ:** 1024

```python
V = 65 * 1024 * 8
a = 200
b = 256
i = int(V / (a*b))
n = 2**i
print (f'Максимальное количество цветов: {n}')
```

**3.** Автоматическая фотокамера с 400 Кбайт видеопамяти производит растровые изображения c фиксированным разрешением и 16-цветной палитрой. Сколько цветов можно будет использовать в палитре, если увеличить видеопамять до 800 Кбайт?

**Ответ:** 256

```python
import math
# V1 = a * b * i1
# V2 = a * b * i2
# V2 / V1 = i2 / i1
# i2 = V2/V1 * i1
# i2 = 2 * i1
i1 = math.log2(16)
i2 = 2 * i1
print (f'Можно будет использовать: {2**i2} цветов')
```

**4.** Автоматическая фотокамера делает фотографии высокого разрешения с палитрой, содержащей 16 777 216 цветов. Средний размер фотографии составляет 12 Мбайт. Для хранения в базе данных фотографии преобразуют в чёрно-белый формат с палитрой, содержащей 256 цветов. Другие преобразования и дополнительные методы сжатия не используются. Сколько Мбайт составляет средний размер преобразованной фотографии?

**Ответ:** 4

```python
import math
# V1 = a*b*i1
# V2 = a*b*i2
# V2 / V1 = i2/i1
# V2 = V1 * i2/i1
i1 = math.log2(16777216)
i2 = math.log2(256)
V1 = 12
V2 = V1 * i2 / i1
print (f'Cредний размер преобразованной фотографии: {V2} Мбайт')
```

**5.** После преобразования растрового 256-цветного графического файла в 4-цветный формат его размер уменьшился на 18 Кбайт. Каков был размер исходного файла в Кбайтах?

**Ответ:** 24

```python
import math
# a*b = size
# V1 = size*i1
# V2 = size*i2
# V1 - V2 = size * (i1 - i2)
# size = (V1 - V2) / (i1 - i2)
i1 = math.log2(256)
i2 = math.log2(4)
size = 18 / (i1 - i2)
V1 = size * i1
print (f'Размер исходного файла: {V1} Кбайт')
```

**6.** Необходимо сохранить изображение размером 960 на 512 пикселей. Известно, что каждый пиксель может быть окрашен в один из 1200 цветов. **8.** Для хранения в информационной системе документы сканируются с разрешением 600 ppi и цветовой системой, содержащей $2^{24}$ = 16 777 216 цветов. Методы сжатия изображений не используются. Средний размер отсканированного документа составляет 16 Мбайт. В целях экономии было решено перейти на разрешение 300 ppi и цветовую систему, содержащую 64 цвета. Сколько Мбайт будет составлять средний размер документа, отсканированного с изменёнными параметрами?

**Ответ:** 1

```python
# V1 = size * ppi1^2 * i1
# V2 = size * ppi2^2 * i2

V1 = 16 * 1024 * 1024 * 8
i1 = 24
size = V1 / (600**2 * i1)

i2 = 6
V2 = size * 300**2 * i2
print (f'средний размер документа, отсканированного с изменёнными параметрами: {V2} бит')
V2 = V2 / 1024 / 1024 / 8
print (f'средний размер документа, отсканированного с изменёнными параметрами: {V2} Мбайт')
```

**1.** Производится двухканальная (стерео) звукозапись с частотой дискретизации 48 кГц. Запись длится 2 минуты 15 секунд, её результаты записываются в файл без сжатия данных, причём каждый сигнал кодируется минимально возможным и одинаковым количеством бит. Информационный объём полученного файла без учета заголовка не превышает 32 Мбайт. Определите максимальную битовую глубину кодирования звука, которая могла быть использована в этой записи.

**Ответ:** 20

```python
# V = k * w * i * t
# i = V / (k*w*t)
V = 32 * 1024 * 1024 * 8
k = 2
w = 48 * 1000
t = 2*60 + 15
i = int(V / (k*w*t))
print (f'Максимальная битовая глубина: {i}')
```

**2.** Производилась двухканальная (стерео) звукозапись с частотой дискретизации 64 кГц и 24-битным разрешением. В результате был получен файл размером 360 Мбайт, сжатие данных не производилось. Определите сколько времени (в минутах) производилась запись. В качестве ответа укажите ближайшее к времени записи целое число минут, кратное 5.

**Ответ:** 15

```python
V = 360 * 1024 * 1024 * 8
k = 2
w = 64 * 1000
i = 24
t = V / (k*w*i)
print (f'Точное время записи: {t} сек')
t = round(t / 60 / 5) * 5
print (f'Приблизительное время записи: {t} мин')
```
