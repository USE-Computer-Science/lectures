**1. Решение через Canvas**

```python
from turtle import *
# отключаем анимацию
tracer(0)
# убираем значок черепахи (лишняя точка)
hideturtle()
# задаём коэффициент масштабирования (достаточно большой для корректного автоподсчёта)
r = 100
# подгоняем размер экрана под размер фигуры
screensize(50*r, 50*r)

# задаём цвета границы (чёрный) и заливки (красный)
color('black', 'red')
# задаём толщину границы заливки (обязательно 1 для автоматического подсчёта)
pensize(1)

up()
# в кумире черепаха смотрит вверх, в python - направо, устраняем неопределённость
left(90)
# смещаем стартовый угол черепахи так, как указано в задании
right(30)
down()
# закрашиваем полученную фигуру
begin_fill()
# отрисовываем фигуру по заданию
for i in range(4):
    forward(25 * r)
    right(90)
end_fill()

# # поднимаем перо черепахи
# up()
# # отрисовываем координатные точки
# for x in range(-50, 50):
#     for y in range(-50, 50):
#         goto(x * r, y * r)
#         dot(3, 'blue')
# # обновляем холст
update()

# получаем отрисованное изображение
canvas = getcanvas()
# счётчик точек в области
cnt = 0
# снова обходим пространство рисунка
for x in range(-300 * r, 300 * r, r):
    for y in range(-300 * r, 300 * r, r):
        # для каждой целой точки получаем список пересечений фигур в ней
        item = canvas.find_overlapping(x,y,x,y)
        # если таких пересечений ровно одно (заливка(5) и фон холста)
        if len(item) == 1 and item[0] == 5:
            print(item)
            cnt += 1

# печатаем ответ
print(cnt)
# завершаем работу с исполнителем черепаха
done()
```

**2. Решение через прямые, построенные с помощью тангенса угла наклона**
