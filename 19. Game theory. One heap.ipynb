#### **19. Задачи с одной кучей камней** 

**Пример 1.** 

Два игрока, Петя и Ваня, играют в следующую игру. Перед игроками лежит куча камней. Игроки ходят по очереди, первый ход делает Петя. За один ход игрок может

а) добавить в кучу один камень;

б) добавить в кучу два камня;

в) увеличить количество камней в куче в два раза;

Игра завершается в тот момент, когда количество камней в куче превышает 54. Победителем считается игрок, сделавший последний ход, то есть первым получивший кучу, в которой будет 55 или больше камней. В начальный момент в куче было S камней, 1 ≤ S ≤ 54.

Ответьте на следующие вопросы:

***Вопрос 1.*** Найдите значение S, при котором Ваня выигрывает своим первым ходом при любой игре Пети.

***Вопрос 2.*** Найдите два значения S, при которых у Пети есть выигрышная стратегия, причём одновременно выполняются два условия:

− Петя не может выиграть за один ход;

− Петя может выиграть своим вторым ходом независимо от того, как будет ходить Ваня.

Найденные значения запишите в ответе через пробел в порядке возрастания.

***Вопрос 3.*** Найдите значение S, при котором одновременно выполняются два условия:

– у Вани есть выигрышная стратегия, позволяющая ему выиграть первым или вторым ходом при любой игре Пети;

– у Вани нет стратегии, которая позволит ему гарантированно выиграть первым ходом.

Ответ:

27

25 26

24


```python
# кэшируем рекурсивную функцию
from functools import lru_cache
@lru_cache(maxsize=None)
def f(s):
  # условие завершения игры
  if s > 54:
    return 0
  # список оценок позиций, в которые можно сходить из текущей позиции
  moves = [f(s+1), f(s+2), f(s*2)]
  # список проигранных для оппонента позиций, в которые можно попасть из текущей позиции
  loses = [x for x in moves if x <= 0]
  # если есть возможность сходить хотя бы в 1 проигрышную позицию
  if loses:
    # текущая позиция - выигрышная за -max(loses) + 1 ходов
    return 1 - max(loses)
  # если же все ходы из текущей позиции ведут в выигрышные позиции
  else:
    # текущая позиция - проигрышная за -max(moves) ходов
    return - max(moves)

# Вопрос 1
for s in range(1,55):
  # f(s) = -1 - позиции, в которых Ваня выигрывает своим первым ходом
  if f(s) == -1:
    print(s)
print('---------')

# Вопрос 2
# заводим список для хранения искомых позиций
wins = []
for s in range(1,55):
  # f(s) = 2 - позиции, в которых Петя выигрывает своим вторым ходом
  if f(s) == 2:
    # добавляем найденные позиции в список
    wins += [s]
# упорядочиваем найденные позиции по возрастанию
print(min(wins), max(wins))
print('---------')

# Вопрос 3
for s in range(1,55):
  # f(s) = -2 - позиции, в которых Ваня выигрывает своим вторым ходом
  if f(s) == -2:
    print(s)
```

**Пример 2.** 

Два игрока, Петя и Ваня, играют в следующую игру. Перед игроками лежит куча камней. Игроки ходят по очереди, первый ход делает Петя. За один ход игрок может добавить в кучу один камень или увеличить количество камней в куче в три раза.

Игра завершается в тот момент, когда количество камней в куче становится не менее 20. Если при этом в куче оказалось не более 40 камней, то победителем считается игрок, сделавший последний ход. В противном случае победителем становится его противник. 

В начальный момент в куче было S камней, 1 ≤ S ≤ 19.
Ответьте на следующие вопросы:

***Вопрос 1.*** Известно, что Ваня выиграл своим первым ходом после неудачного первого хода Пети. Назовите минимальное значение S, при котором это возможно.

***Вопрос 2.*** Определите два таких значения S, при которых у Пети есть выигрышная стратегия, причём одновременно выполняются два условия:

− Петя не может выиграть за один ход;

− Петя может выиграть своим вторым ходом независимо от того, как будет ходить Ваня.

Найденные значения запишите в ответе через пробел в порядке возрастания.

***Вопрос 3.*** Найдите значение S, при котором одновременно выполняются два условия:

– у Вани есть выигрышная стратегия, позволяющая ему выиграть первым или вторым ходом при любой игре Пети;

– у Вани нет стратегии, которая позволит ему гарантированно выиграть первым ходом.


Ответ

1) 3

2) 6 17

3) 16

```python
# кэшируем рекурсивную функцию
from functools import lru_cache
@lru_cache(maxsize=None)
def f(s):
  # условие победного завершения игры
  if s >= 20 and s <= 40:
    return 0
  # условие проигрыша
  # мы хотим, чтобы игрок сходивший в проигранную позицию имел позицию проигрышную в 1 ход
  # следовательно уже проигранная позиция должна оцениваться, как выигранная в 1 ход (как будто второй игрок сходил из неё и выиграл)  
  if s > 40:
    return 1
  # список оценок позиций, в которые можно сходить из текущей позиции
  moves = [f(s+1), f(s*3)]
  # список проигранных для оппонента позиций, в которые можно попасть из текущей позиции
  loses = [x for x in moves if x <= 0]
  # если есть возможность сходить хотя бы в 1 проигрышную позицию
  if loses:
    # текущая позиция - выигрышная за -max(loses) + 1 ходов
    return 1 - max(loses)
  # если же все ходы из текущей позиции ведут в выигрышные позиции
  else:
    # текущая позиция - проигрышная за -max(moves) ходов
    return - max(moves)

# Вопрос 1
for s in range(1,20):
  # Петя создал для Вани выигрышную в 1 ход позицию (f(s) = 1)
  # он мог сделать это либо добавив 1 камень в кучу, либо увеличив количество камней в куче в 3 раза
  # найдём минимальное s, при котором это возможно
  if f(s+1) == 1 or f(s*3) == 1:
    print(s)
    # т.к. нас интересует минимальное s - можем завершить поиск после нахождения первого подходящего s
    break

# Вопрос 2
# заводим список для хранения искомых позиций
wins = []
for s in range(1,20):
  # f(s) = 2 - позиции, в которых Петя выигрывает своим вторым ходом
  if f(s) == 2:
    # добавляем найденные позиции в список
    wins += [s]
# упорядочиваем найденные позиции по возрастанию
print(min(wins), max(wins))

# Вопрос 3
wins = []
for s in range(1,20):
  # f(s) = -2 - позиции, в которых Ваня выигрывает своим вторым ходом
  if f(s) == -2:
    print(s)
```

**Пример 3.** 

Два игрока, Паша и Витя, играют в следующую игру. Перед игроками лежит куча камней. Игроки ходят по очереди, первый ход делает Паша. За один ход игрок может убрать из кучи один камень или уменьшить количество камней в куче в два раза (если количество камней в куче нечётно, остаётся на 1 камень больше, чем убирается). Например, имея кучу из 15 камней, за один ход можно получить кучу из 14 или 8 камней. Игра завершается в тот момент, когда количество камней в куче становится не более 33. Победителем считается игрок, сделавший последний ход, то есть первым получивший позицию, при которой в куче будет 33 или меньше камней. В начальный момент в куче было S камней, S > 33. 
Ответьте на следующие вопросы:

***Вопрос 1.*** Найдите значение S, при котором Витя выигрывает своим первым ходом при любой игре Паши.

***Вопрос 2.*** Найдите минимальное и максимальное значения S, при которых у Паши есть выигрышная стратегия, причём одновременно выполняются два условия: 

− Паша не может выиграть за один ход; 

− Паша может выиграть своим вторым ходом независимо от того, как будет ходить Витя. 

Найденные значения запишите в ответе через пробел в порядке возрастания.

***Вопрос 3.*** Найдите значения S, при которых одновременно выполняются два условия: 

– у Вити есть выигрышная стратегия, позволяющая ему выиграть первым или вторым ходом при любой игре Паши; 

– у Вити нет стратегии, которая позволит ему гарантированно выиграть первым ходом.

Найденные значения запишите в ответе через пробел в порядке возрастания.


Ответ:

1) 67

2) 68 143

3) 69 135

```python
# кэшируем рекурсивную функцию
from functools import lru_cache
@lru_cache(maxsize=None)
def f(s):
  # условие завершения игры
  if s <= 33:
    return 0
  # список оценок позиций, в которые можно сходить из текущей позиции
  # один камень можно убрать всегда
  moves = [f(s-1)]
  # если текущее количество камней в куче чётно
  if s % 2 == 0:
    # можно убрать половину камней
    moves += [f(s//2)]
  # если текущее количество камней в куче нечётно
  else:
    # можно убрать на один камень меньше, чем останется в куче
    moves += [f(s//2 + 1)]
  # список проигранных для оппонента позиций, в которые можно попасть из текущей позиции
  loses = [x for x in moves if x <= 0]
  # если есть возможность сходить хотя бы в 1 проигрышную позицию
  if loses:
    # текущая позиция - выигрышная за -max(loses) + 1 ходов
    return 1 - max(loses)
  # если же все ходы из текущей позиции ведут в выигрышные позиции
  else:
    # текущая позиция - проигрышная за -max(moves) ходов
    return - max(moves)

# Вопрос 1
for s in range(33,100):
  # f(s) = -1 - позиции, в которых Витя выигрывает своим первым ходом
  if f(s) == -1:
    print(s)

# Вопрос 2
# заводим список для хранения искомых позиций
wins = []
for s in range(33,200):
  # f(s) = 2 - позиции, в которых Паша выигрывает своим вторым ходом
  if f(s) == 2:
    # добавляем найденные позиции в список
    wins += [s]
# выводим минимальное и максимальное найденные значения s
print(min(wins), max(wins))

# Вопрос 3
# заводим список для хранения искомых позиций
wins = []
for s in range(33,200):
  # f(s) = -2 - позиции, в которых Витя выигрывает своим вторым ходом
  if f(s) == -2:
    # добавляем найденные позиции в список
    wins += [s]
# упорядочиваем найденные позиции по возрастанию
print(wins[0], wins[1])
```
