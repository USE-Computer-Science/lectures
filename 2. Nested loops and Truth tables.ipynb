Клетки шахматной доски индексируются цифрами от 1 до 8 слева направо и сверху вниз. 
image.png
<br>Напишите функцию, которая по заданным начальным координатам ладьи (слона, коня) возвращает список клеток, в которые может переместиться эта фигура за 1 ход. 

```python
# функция, возвращающая список возможных ходов ладьи
# get_rook_moves(x_ладьи, y_ладьи) -> list: [(1,1), ... , (8,1)]
def get_rook_moves(x, y):
  lst =[]
  # обходим все клетки по горизонтали
  for i in range(1,9):
    # обходим все клетки по вертикали
    for j in range(1,9):
      # находим клетку, x или y координата которой совпадает с соответствующей текущей координатой ладьи
      if x == i or j == y:
        # исключаем клетку, в которой сейчас находится ладья
        if x == i and j == y:
          continue
        # добавляем найденную клетку
        lst.append((i,j))
  # возвращаем список найденных клеток
  return lst

# находим ходы ладьи из клетки (2,2)
get_rook_moves(2, 2)
```

```python
# функция, возвращающая список возможных ходов слона
# get_bishop_moves(x_слона, y_слона) -> list: [(1,1), ... , (8,8)]
def get_bishop_moves(x, y):
  lst =[]
  # обходим все клетки по горизонтали
  for i in range(1,9):
    # обходим все клетки по вертикали
    for j in range(1,9):
      # находим клетку, равноудалённую по x и по y от начальной клетки слона (т.е. клетки на одних с ним диагоналях)
      if abs(x - i) == abs(y - j):
        # исключаем клетку, в которой сейчас находится слон
        if x == i and j == y:
          continue
        # добавляем найденную клетку
        lst.append((i,j))
  # возвращаем список найденных клеток
  return lst

# находим ходы слона из клетки (6,4)
get_bishop_moves(6, 4)
```

```python
# функция, возвращающая список возможных ходов коня
# get_knight_moves(x_коня, y_коня) -> list: [(1,1), ... , (3,2)]
def get_knight_moves(x, y):
  lst =[]
  # обходим все клетки по горизонтали
  for i in range(1,9):
    # обходим все клетки по вертикали
    for j in range(1,9):
      # координата коня всегда меняется на 2 по одной из осей, и на 1 по другой
      # находим клетку, координаты которой отличаются от текущей на 2 по x, и на 1 y или наоборот
      if (abs(x - i) == 1 and abs(y - j) == 2) or (abs(x - i) == 2 and abs(y - j) == 1):
        # добавляем найденную клетку
        lst.append((i,j))
  # возвращаем список найденных клеток
  return lst

# находим ходы коня из клетки (6,8)
get_knight_moves(6, 8)
```

Напишите функцию, которая по заданным координатам ладьи и слона возвращает список клеток, которые находятся под боем обеих этих фигур

```python
# функция, возвращающая список возможных ходов для ладьи и слона одновременно
# get_double_attacked_squares(x_ладьи, y_ладьи, x_слона, y_слона) -> list: [(2,8), ... , (8,2)]
def get_double_attacked_squares(rook_x, rook_y, bishop_x, bishop_y):
  lst = []
  # находим ходы ладьи из клетки (x_ладьи, y_ладьи)
  lst_rook = get_rook_moves(rook_x, rook_y)
  # находим ходы слона из клетки (x_слона, y_слона)
  lst_bishop = get_bishop_moves(bishop_x, bishop_y)
  # обходим список ходов ладьи
  for item_r in lst_rook:
    # обходим список ходов слона
    for item_b in lst_bishop:
      # если ходы совпадают
      if item_r == item_b:
        # добавляем в список возможных ходов для обеих фигур
        lst.append(item_r)
  # возвращаем список найденных клеток
  return lst
  # можно и через пересечение множеств ходов двух фигур
  #return list( set(lst_rook) & set(lst_bishop) )

# находим совместные ходы ладьи из клетки (2,2) и слона из клетки (6,8)
get_double_attacked_squares(2,2,6,4)
```

**3470** Логическая функция F задаётся выражением F = (x ∧ (y ∨ ¬z) ∧ w) ≡ (x → ¬y ∧ z). <br>Напечатайте значения переменных, на которых эта функция истинна

```python
print('x y z w')         # печатаем шапку
for x in range(2):       # перебираем все значения переменной x
  for y in range(2):     # перебираем все значения переменной y
    for z in range(2):   # перебираем все значения переменной z
      for w in range(2): # перебираем все значения переменной w
        # создаём программное представление функции F
        # очень сильно не хватало дополнительных скобок при раскрытии импликации: (x → ¬y ∧ z) =>  (x <= ((not y) and z))
        F = (x and (y or (not z)) and w) == (x <= ((not y) and z))
        if F:
          print(x,y,z,w)
```
